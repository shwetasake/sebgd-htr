```python
# ============================================================
# ICDAR HWD 2024 — WORD LEVEL OCR (CORRECTED VERSION)
# Stable Attention OCR Training
# ============================================================

import os
import random
import numpy as np
from PIL import Image

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms

from tqdm import tqdm
import matplotlib.pyplot as plt

# ============================================================
# DEVICE
# ============================================================

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

print(f"Device : {device}")

if torch.cuda.is_available():
    print(f"GPU    : {torch.cuda.get_device_name(0)}")
    print(f"VRAM   : {torch.cuda.get_device_properties(0).total_memory/1e9:.1f} GB")

# ============================================================
# CONFIG
# ============================================================

DATASET_ROOT = "/home/mca/Shweta/handwritten text detection/dataset/Word_Level_English_Training_Set/Word_Level_Training_Set"

GT_FILE = "train.txt"

SAVE_DIR = "/home/mca/Shweta/handwritten text detection/saved_models/icdar_word_corrected"

os.makedirs(SAVE_DIR, exist_ok=True)

BATCH_SIZE = 32
NUM_EPOCHS = 30
LEARNING_RATE = 1e-4
NUM_WORKERS = 4

MAX_LABEL_LEN = 25

IMG_H = 32
IMG_W = 160

VAL_SPLIT = 0.1

PATIENCE = 7

HIDDEN_SIZE = 256

# ============================================================
# VOCAB
# ============================================================

CHARS = ' !"#$%&\'()*+,-./0123456789:;<=>?@ABCDEFGHIJKLMNOPQRSTUVWXYZ[\\]^_`abcdefghijklmnopqrstuvwxyz{|}~'

GO_IDX  = 0
EOS_IDX = 1
PAD_IDX = 2
UNK_IDX = 3

CHAR_TO_IDX = {c: i + 4 for i, c in enumerate(CHARS)}
IDX_TO_CHAR = {v: k for k, v in CHAR_TO_IDX.items()}

IDX_TO_CHAR.update({
    GO_IDX: "<GO>",
    EOS_IDX: "<EOS>",
    PAD_IDX: "<PAD>",
    UNK_IDX: "<UNK>"
})

NUM_CLASSES = len(CHARS) + 4

print("NUM_CLASSES =", NUM_CLASSES)

# ============================================================
# LABEL ENCODING
# ============================================================

def encode_label(text, max_len):

    tokens = [GO_IDX]

    for c in text[:max_len]:
        tokens.append(CHAR_TO_IDX.get(c, UNK_IDX))

    tokens.append(EOS_IDX)

    while len(tokens) < max_len + 2:
        tokens.append(PAD_IDX)

    return torch.tensor(tokens[:max_len+2], dtype=torch.long)


def decode_label(tensor):

    chars = []

    for idx in tensor.tolist():

        if idx in [EOS_IDX, PAD_IDX]:
            break

        if idx in [GO_IDX, UNK_IDX]:
            continue

        chars.append(IDX_TO_CHAR.get(idx, ""))

    return "".join(chars)

# ============================================================
# TRANSFORMS
# ============================================================

def get_transforms(imgH, imgW, augment=False):

    ops = [
        transforms.Grayscale(1),
        transforms.Resize((imgH, imgW))
    ]

    if augment:
        ops += [
            transforms.RandomAffine(
                degrees=2,
                translate=(0.02, 0.02),
                scale=(0.95, 1.05),
                shear=2
            )
        ]

    ops += [
        transforms.ToTensor(),
        transforms.Normalize([0.5], [0.5])
    ]

    return transforms.Compose(ops)

# ============================================================
# DATASET
# ============================================================

def load_gt_file(gt_path, dataset_root, max_label_len):

    samples = []

    with open(gt_path, 'r', encoding='utf-8') as f:

        for line in f:

            line = line.rstrip('\n')

            if '\t' not in line:
                continue

            rel_path, label = line.split('\t', 1)

            label = label.strip()

            if len(label) == 0:
                continue

            if len(label) > max_label_len:
                continue

            abs_path = os.path.join(dataset_root, rel_path)

            if not os.path.exists(abs_path):
                continue

            samples.append((abs_path, label))

    print(f"Loaded samples : {len(samples)}")

    return samples


class ICDARWordDataset(Dataset):

    def __init__(
        self,
        samples,
        max_label_len=25,
        imgH=32,
        imgW=160,
        augment=False
    ):

        self.samples = samples
        self.max_label_len = max_label_len

        self.transform = get_transforms(
            imgH,
            imgW,
            augment
        )

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):

        img_path, label = self.samples[idx]

        try:
            image = Image.open(img_path).convert("RGB")

        except:
            image = Image.new("RGB", (IMG_W, IMG_H), (255,255,255))

        image = self.transform(image)

        encoded = encode_label(
            label,
            self.max_label_len
        )

        return image, encoded, label

# ============================================================
# SPLIT
# ============================================================

def make_train_val_split(samples, val_fraction=0.1):

    samples = samples.copy()

    random.seed(42)

    random.shuffle(samples)

    n_val = int(len(samples) * val_fraction)

    val_samples = samples[:n_val]
    train_samples = samples[n_val:]

    print("Train =", len(train_samples))
    print("Val   =", len(val_samples))

    return train_samples, val_samples

# ============================================================
# CNN FEATURE EXTRACTOR
# ============================================================

class CNNFeatureExtractor(nn.Module):

    def __init__(self):

        super().__init__()

        self.cnn = nn.Sequential(

            nn.Conv2d(1, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2,2),

            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2,2),

            nn.Conv2d(128, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),

            nn.Conv2d(256, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),

            nn.MaxPool2d((2,1), (2,1)),

            nn.Conv2d(256, 512, 3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True),

            nn.Conv2d(512, 512, 3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True),

            nn.MaxPool2d((2,1), (2,1)),

            nn.Conv2d(512, 512, 2),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):

        return self.cnn(x)

# ============================================================
# BiLSTM
# ============================================================

class BidirectionalLSTM(nn.Module):

    def __init__(
        self,
        input_size,
        hidden_size,
        output_size
    ):

        super().__init__()

        self.rnn = nn.LSTM(
            input_size,
            hidden_size,
            bidirectional=True,
            batch_first=True
        )

        self.linear = nn.Linear(
            hidden_size * 2,
            output_size
        )

    def forward(self, x):

        x, _ = self.rnn(x)

        x = self.linear(x)

        return x

# ============================================================
# ATTENTION
# ============================================================

class AttentionCell(nn.Module):

    def __init__(
        self,
        input_size,
        hidden_size,
        num_embeddings
    ):

        super().__init__()

        self.i2h = nn.Linear(
            input_size,
            hidden_size,
            bias=False
        )

        self.h2h = nn.Linear(
            hidden_size,
            hidden_size
        )

        self.score = nn.Linear(
            hidden_size,
            1,
            bias=False
        )

        self.rnn = nn.LSTMCell(
            input_size + num_embeddings,
            hidden_size
        )

    def forward(
        self,
        prev_hidden,
        batch_H,
        char_onehots
    ):

        batch_H_proj = self.i2h(batch_H)

        prev_hidden_proj = self.h2h(prev_hidden[0]).unsqueeze(1)

        e = self.score(
            torch.tanh(batch_H_proj + prev_hidden_proj)
        )

        alpha = F.softmax(e, dim=1)

        context = torch.bmm(
            alpha.permute(0,2,1),
            batch_H
        ).squeeze(1)

        concat_context = torch.cat(
            [context, char_onehots],
            dim=1
        )

        cur_hidden = self.rnn(
            concat_context,
            prev_hidden
        )

        return cur_hidden, alpha

# ============================================================
# ATTENTION DECODER
# ============================================================

class Attention(nn.Module):

    def __init__(
        self,
        input_size,
        hidden_size,
        num_classes
    ):

        super().__init__()

        self.attention_cell = AttentionCell(
            input_size,
            hidden_size,
            num_classes
        )

        self.hidden_size = hidden_size
        self.num_classes = num_classes

        self.generator = nn.Linear(
            hidden_size,
            num_classes
        )

    def _char_to_onehot(self, input_char):

        one_hot = torch.zeros(
            input_char.size(0),
            self.num_classes,
            device=input_char.device
        )

        one_hot.scatter_(1, input_char.unsqueeze(1), 1)

        return one_hot

    def forward(
        self,
        batch_H,
        text,
        is_train=True,
        batch_max_length=25
    ):

        batch_size = batch_H.size(0)

        num_steps = batch_max_length + 1

        hidden = (
            torch.zeros(batch_size, self.hidden_size, device=batch_H.device),
            torch.zeros(batch_size, self.hidden_size, device=batch_H.device)
        )

        output_hiddens = torch.zeros(
            batch_size,
            num_steps,
            self.hidden_size,
            device=batch_H.device
        )

        # ====================================================
        # TRAIN
        # ====================================================

        if is_train:

            for i in range(num_steps):

                char_onehots = self._char_to_onehot(text[:, i])

                hidden, alpha = self.attention_cell(
                    hidden,
                    batch_H,
                    char_onehots
                )

                output_hiddens[:, i, :] = hidden[0]

            probs = self.generator(output_hiddens)

        # ====================================================
        # INFERENCE
        # ====================================================

        else:

            targets = torch.LongTensor(
                batch_size
            ).fill_(GO_IDX).to(batch_H.device)

            probs = torch.zeros(
                batch_size,
                num_steps,
                self.num_classes,
                device=batch_H.device
            )

            for i in range(num_steps):

                char_onehots = self._char_to_onehot(targets)

                hidden, alpha = self.attention_cell(
                    hidden,
                    batch_H,
                    char_onehots
                )

                probs_step = self.generator(hidden[0])

                probs[:, i, :] = probs_step

                _, next_input = probs_step.max(1)

                targets = next_input

        return probs

# ============================================================
# FULL OCR MODEL
# ============================================================

class OCRModel(nn.Module):

    def __init__(self):

        super().__init__()

        # TPS REMOVED
        self.Transformation = nn.Identity()

        self.FeatureExtraction = CNNFeatureExtractor()

        self.AdaptiveAvgPool = nn.AdaptiveAvgPool2d((None,1))

        self.SequenceModeling = nn.Sequential(

            BidirectionalLSTM(
                512,
                HIDDEN_SIZE,
                HIDDEN_SIZE
            ),

            BidirectionalLSTM(
                HIDDEN_SIZE,
                HIDDEN_SIZE,
                HIDDEN_SIZE
            )
        )

        self.Prediction = Attention(
            HIDDEN_SIZE,
            HIDDEN_SIZE,
            NUM_CLASSES
        )

    def forward(
        self,
        images,
        text,
        is_train=True
    ):

        visual_feature = self.FeatureExtraction(images)

        visual_feature = self.AdaptiveAvgPool(
            visual_feature.permute(0,3,1,2)
        )

        visual_feature = visual_feature.squeeze(3)

        contextual_feature = self.SequenceModeling(
            visual_feature
        )

        prediction = self.Prediction(
            contextual_feature.contiguous(),
            text,
            is_train,
            MAX_LABEL_LEN
        )

        return prediction

# ============================================================
# LOAD DATA
# ============================================================

gt_path = os.path.join(
    DATASET_ROOT,
    GT_FILE
)

all_samples = load_gt_file(
    gt_path,
    DATASET_ROOT,
    MAX_LABEL_LEN
)

train_samples, val_samples = make_train_val_split(
    all_samples,
    VAL_SPLIT
)

train_ds = ICDARWordDataset(
    train_samples,
    MAX_LABEL_LEN,
    IMG_H,
    IMG_W,
    augment=True
)

val_ds = ICDARWordDataset(
    val_samples,
    MAX_LABEL_LEN,
    IMG_H,
    IMG_W,
    augment=False
)

train_loader = DataLoader(
    train_ds,
    batch_size=BATCH_SIZE,
    shuffle=True,
    num_workers=NUM_WORKERS,
    pin_memory=True
)

val_loader = DataLoader(
    val_ds,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=NUM_WORKERS,
    pin_memory=True
)

# ============================================================
# MODEL
# ============================================================

model = OCRModel().to(device)

print("Parameters =", sum(
    p.numel()
    for p in model.parameters()
    if p.requires_grad
))

# ============================================================
# LOSS
# ============================================================

criterion = nn.CrossEntropyLoss(
    ignore_index=PAD_IDX,
    label_smoothing=0.1
)

optimizer = optim.Adam(
    model.parameters(),
    lr=LEARNING_RATE
)

scheduler = optim.lr_scheduler.CosineAnnealingLR(
    optimizer,
    T_max=NUM_EPOCHS
)

# ============================================================
# METRICS
# ============================================================

def compute_metrics(logits, labels):

    preds = logits.argmax(dim=-1)

    total_words = 0
    correct_words = 0

    total_chars = 0
    correct_chars = 0

    for pred_seq, gt_seq in zip(preds, labels):

        pred_str = decode_label(pred_seq)
        gt_str = decode_label(gt_seq)

        total_words += 1

        if pred_str == gt_str:
            correct_words += 1

        for p, g in zip(pred_str, gt_str):

            total_chars += 1

            if p == g:
                correct_chars += 1

    word_acc = correct_words / total_words

    char_acc = correct_chars / max(total_chars, 1)

    return word_acc, char_acc

# ============================================================
# VALIDATION
# ============================================================

def run_validation():

    model.eval()

    total_loss = 0

    total_wacc = 0
    total_cacc = 0

    n = 0

    with torch.no_grad():

        for images, labels, _ in val_loader:

            images = images.to(device)
            labels = labels.to(device)

            dummy_text = torch.zeros(
                labels.size(0),
                MAX_LABEL_LEN + 1,
                dtype=torch.long,
                device=device
            )

            logits = model(
                images,
                dummy_text,
                is_train=False
            )

            target = labels[:, 1:]

            loss = criterion(
                logits.reshape(-1, NUM_CLASSES),
                target.reshape(-1)
            )

            wacc, cacc = compute_metrics(
                logits.cpu(),
                labels.cpu()
            )

            total_loss += loss.item()

            total_wacc += wacc
            total_cacc += cacc

            n += 1

    return (
        total_loss / n,
        total_wacc / n,
        total_cacc / n
    )

# ============================================================
# TRAIN
# ============================================================

best_val_acc = 0

history = {
    "train_loss": [],
    "val_loss": [],
    "val_wacc": []
}

for epoch in range(NUM_EPOCHS):

    model.train()

    running_loss = 0

    pbar = tqdm(train_loader)

    for images, labels, _ in pbar:

        images = images.to(device)
        labels = labels.to(device)

        logits = model(
            images,
            labels[:, :-1],
            is_train=True
        )

        target = labels[:, 1:]

        loss = criterion(
            logits.reshape(-1, NUM_CLASSES),
            target.reshape(-1)
        )

        optimizer.zero_grad()

        loss.backward()

        torch.nn.utils.clip_grad_norm_(
            model.parameters(),
            5.0
        )

        optimizer.step()

        running_loss += loss.item()

        pbar.set_description(
            f"Epoch {epoch+1}/{NUM_EPOCHS}"
        )

        pbar.set_postfix(
            loss=f"{loss.item():.4f}"
        )

    scheduler.step()

    train_loss = running_loss / len(train_loader)

    val_loss, val_wacc, val_cacc = run_validation()

    history["train_loss"].append(train_loss)
    history["val_loss"].append(val_loss)
    history["val_wacc"].append(val_wacc)

    print()

    print(f"Epoch {epoch+1}")
    print(f"Train Loss : {train_loss:.4f}")
    print(f"Val Loss   : {val_loss:.4f}")
    print(f"Val WAcc   : {val_wacc*100:.2f}%")
    print(f"Val CAcc   : {val_cacc*100:.2f}%")

    torch.save(
        model.state_dict(),
        os.path.join(SAVE_DIR, "last.pth")
    )

    if val_wacc > best_val_acc:

        best_val_acc = val_wacc

        torch.save(
            model.state_dict(),
            os.path.join(SAVE_DIR, "best.pth")
        )

        print("BEST MODEL SAVED")

print()

print("Training Complete")

print(f"Best Val Accuracy : {best_val_acc*100:.2f}%")

# ============================================================
# PLOT
# ============================================================

plt.figure(figsize=(10,5))

plt.plot(history["train_loss"], label="Train Loss")
plt.plot(history["val_loss"], label="Val Loss")

plt.legend()

plt.grid(True)

plt.savefig(
    os.path.join(SAVE_DIR, "loss_curve.png")
)

plt.show()
```

```python
# ============================================================
# IAM WORD OCR — TRAIN / VAL / TEST SPLIT
# CORRECTED COMPLETE TRAINING CODE
# ============================================================

import os
import random
import numpy as np
from PIL import Image

import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms

from tqdm import tqdm
import matplotlib.pyplot as plt

# ============================================================
# DEVICE
# ============================================================

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

print(f"Device : {device}")

if torch.cuda.is_available():
    print(f"GPU    : {torch.cuda.get_device_name(0)}")
    print(f"VRAM   : {torch.cuda.get_device_properties(0).total_memory/1e9:.1f} GB")

# ============================================================
# CONFIG
# ============================================================

DATASET_ROOT = "/home/mca/Shweta/handwritten text detection/dataset/iam/iam_words"

GT_FILE = os.path.join(DATASET_ROOT, "words.txt")

SAVE_DIR = "./saved_iam_2"

os.makedirs(SAVE_DIR, exist_ok=True)

BATCH_SIZE = 32
NUM_EPOCHS = 30
LEARNING_RATE = 1e-4
NUM_WORKERS = 4

IMG_H = 32
IMG_W = 160

MAX_LABEL_LEN = 25

TRAIN_SPLIT = 0.8
VAL_SPLIT   = 0.1
TEST_SPLIT  = 0.1

PATIENCE = 7

HIDDEN_SIZE = 256

SEED = 42

# ============================================================
# FIX SEED
# ============================================================

random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)

if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

# ============================================================
# VOCAB
# ============================================================

CHARS = ' !"#$%&\'()*+,-./0123456789:;<=>?@ABCDEFGHIJKLMNOPQRSTUVWXYZ[\\]^_`abcdefghijklmnopqrstuvwxyz{|}~'

GO_IDX  = 0
EOS_IDX = 1
PAD_IDX = 2
UNK_IDX = 3

CHAR_TO_IDX = {c: i + 4 for i, c in enumerate(CHARS)}
IDX_TO_CHAR = {v: k for k, v in CHAR_TO_IDX.items()}

IDX_TO_CHAR.update({
    GO_IDX: "<GO>",
    EOS_IDX: "<EOS>",
    PAD_IDX: "<PAD>",
    UNK_IDX: "<UNK>"
})

NUM_CLASSES = len(CHARS) + 4

print("NUM_CLASSES =", NUM_CLASSES)

# ============================================================
# LABEL ENCODING
# ============================================================

def encode_label(text, max_len):

    tokens = [GO_IDX]

    for c in text[:max_len]:
        tokens.append(CHAR_TO_IDX.get(c, UNK_IDX))

    tokens.append(EOS_IDX)

    while len(tokens) < max_len + 2:
        tokens.append(PAD_IDX)

    return torch.tensor(tokens[:max_len+2], dtype=torch.long)


def decode_label(tensor):

    chars = []

    for idx in tensor.tolist():

        if idx == EOS_IDX:
            break

        if idx in [GO_IDX, PAD_IDX, UNK_IDX]:
            continue

        chars.append(IDX_TO_CHAR.get(idx, ""))

    return "".join(chars)

# ============================================================
# TRANSFORMS
# ============================================================

def get_transforms(imgH, imgW, augment=False):

    ops = [
        transforms.Grayscale(1),
        transforms.Resize((imgH, imgW))
    ]

    if augment:
        ops += [
            transforms.RandomAffine(
                degrees=2,
                translate=(0.02, 0.02),
                scale=(0.95, 1.05),
                shear=2
            )
        ]

    ops += [
        transforms.ToTensor(),
        transforms.Normalize([0.5], [0.5])
    ]

    return transforms.Compose(ops)

# ============================================================
# LOAD IAM WORDS.TXT
# ============================================================

def load_iam_words(gt_file, dataset_root, max_label_len):

    samples = []

    skipped = 0

    with open(gt_file, "r", encoding="utf-8") as f:

        for line in f:

            line = line.strip()

            if line.startswith("#"):
                continue

            if len(line) == 0:
                continue

            parts = line.split()

            if len(parts) < 9:
                skipped += 1
                continue

            word_id = parts[0]

            status = parts[1]

            # skip bad segmentation
            if status != "ok":
                skipped += 1
                continue

            label = " ".join(parts[8:])

            if len(label) == 0:
                skipped += 1
                continue

            if len(label) > max_label_len:
                skipped += 1
                continue

            # ------------------------------------------------
            # PATH CREATION
            #
            # a01-000u-00-00
            # ->
            # words/a01/a01-000u/a01-000u-00-00.png
            # ------------------------------------------------

            folder1 = word_id.split("-")[0]

            folder2 = "-".join(word_id.split("-")[:2])

            img_path = os.path.join(
                dataset_root,
                "words",
                folder1,
                folder2,
                word_id + ".png"
            )

            if not os.path.exists(img_path):
                skipped += 1
                continue

            samples.append((img_path, label))

    print(f"Loaded samples : {len(samples)}")
    print(f"Skipped        : {skipped}")

    return samples

# ============================================================
# DATASET
# ============================================================

class IAMDataset(Dataset):

    def __init__(
        self,
        samples,
        imgH=32,
        imgW=160,
        max_label_len=25,
        augment=False
    ):

        self.samples = samples

        self.max_label_len = max_label_len

        self.transform = get_transforms(
            imgH,
            imgW,
            augment
        )

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):

        img_path, label = self.samples[idx]

        try:
            image = Image.open(img_path).convert("RGB")

        except:
            image = Image.new("RGB", (IMG_W, IMG_H), (255,255,255))

        image = self.transform(image)

        encoded = encode_label(
            label,
            self.max_label_len
        )

        return image, encoded, label

# ============================================================
# TRAIN / VAL / TEST SPLIT
# ============================================================

def make_splits(samples):

    random.shuffle(samples)

    total = len(samples)

    n_train = int(total * TRAIN_SPLIT)
    n_val   = int(total * VAL_SPLIT)

    train_samples = samples[:n_train]

    val_samples = samples[n_train:n_train+n_val]

    test_samples = samples[n_train+n_val:]

    print()
    print("Dataset Split")
    print("---------------------------")
    print("Train :", len(train_samples))
    print("Val   :", len(val_samples))
    print("Test  :", len(test_samples))

    return train_samples, val_samples, test_samples

# ============================================================
# CNN FEATURE EXTRACTOR
# ============================================================

class CNNFeatureExtractor(nn.Module):

    def __init__(self):

        super().__init__()

        self.cnn = nn.Sequential(

            nn.Conv2d(1, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2,2),

            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2,2),

            nn.Conv2d(128, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),

            nn.Conv2d(256, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),

            nn.MaxPool2d((2,1), (2,1)),

            nn.Conv2d(256, 512, 3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True),

            nn.Conv2d(512, 512, 3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True),

            nn.MaxPool2d((2,1), (2,1)),

            nn.Conv2d(512, 512, 2),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):

        return self.cnn(x)

# ============================================================
# BiLSTM
# ============================================================

class BidirectionalLSTM(nn.Module):

    def __init__(
        self,
        input_size,
        hidden_size,
        output_size
    ):

        super().__init__()

        self.rnn = nn.LSTM(
            input_size,
            hidden_size,
            bidirectional=True,
            batch_first=True
        )

        self.linear = nn.Linear(
            hidden_size * 2,
            output_size
        )

    def forward(self, x):

        x, _ = self.rnn(x)

        x = self.linear(x)

        return x

# ============================================================
# ATTENTION
# ============================================================

class AttentionCell(nn.Module):

    def __init__(
        self,
        input_size,
        hidden_size,
        num_embeddings
    ):

        super().__init__()

        self.i2h = nn.Linear(
            input_size,
            hidden_size,
            bias=False
        )

        self.h2h = nn.Linear(
            hidden_size,
            hidden_size
        )

        self.score = nn.Linear(
            hidden_size,
            1,
            bias=False
        )

        self.rnn = nn.LSTMCell(
            input_size + num_embeddings,
            hidden_size
        )

    def forward(
        self,
        prev_hidden,
        batch_H,
        char_onehots
    ):

        batch_H_proj = self.i2h(batch_H)

        prev_hidden_proj = self.h2h(prev_hidden[0]).unsqueeze(1)

        e = self.score(
            torch.tanh(batch_H_proj + prev_hidden_proj)
        )

        alpha = F.softmax(e, dim=1)

        context = torch.bmm(
            alpha.permute(0,2,1),
            batch_H
        ).squeeze(1)

        concat_context = torch.cat(
            [context, char_onehots],
            dim=1
        )

        cur_hidden = self.rnn(
            concat_context,
            prev_hidden
        )

        return cur_hidden, alpha

# ============================================================
# ATTENTION DECODER
# ============================================================

class Attention(nn.Module):

    def __init__(
        self,
        input_size,
        hidden_size,
        num_classes
    ):

        super().__init__()

        self.attention_cell = AttentionCell(
            input_size,
            hidden_size,
            num_classes
        )

        self.hidden_size = hidden_size
        self.num_classes = num_classes

        self.generator = nn.Linear(
            hidden_size,
            num_classes
        )

    def _char_to_onehot(self, input_char):

        one_hot = torch.zeros(
            input_char.size(0),
            self.num_classes,
            device=input_char.device
        )

        one_hot.scatter_(1, input_char.unsqueeze(1), 1)

        return one_hot

    def forward(
        self,
        batch_H,
        text,
        is_train=True,
        batch_max_length=25
    ):

        batch_size = batch_H.size(0)

        num_steps = batch_max_length + 1

        hidden = (
            torch.zeros(batch_size, self.hidden_size, device=batch_H.device),
            torch.zeros(batch_size, self.hidden_size, device=batch_H.device)
        )

        output_hiddens = torch.zeros(
            batch_size,
            num_steps,
            self.hidden_size,
            device=batch_H.device
        )

        if is_train:

            for i in range(num_steps):

                char_onehots = self._char_to_onehot(text[:, i])

                hidden, alpha = self.attention_cell(
                    hidden,
                    batch_H,
                    char_onehots
                )

                output_hiddens[:, i, :] = hidden[0]

            probs = self.generator(output_hiddens)

        else:

            targets = torch.LongTensor(
                batch_size
            ).fill_(GO_IDX).to(batch_H.device)

            probs = torch.zeros(
                batch_size,
                num_steps,
                self.num_classes,
                device=batch_H.device
            )

            for i in range(num_steps):

                char_onehots = self._char_to_onehot(targets)

                hidden, alpha = self.attention_cell(
                    hidden,
                    batch_H,
                    char_onehots
                )

                probs_step = self.generator(hidden[0])

                probs[:, i, :] = probs_step

                _, next_input = probs_step.max(1)

                targets = next_input

        return probs

# ============================================================
# FULL OCR MODEL
# ============================================================

class OCRModel(nn.Module):

    def __init__(self):

        super().__init__()

        self.FeatureExtraction = CNNFeatureExtractor()

        self.AdaptiveAvgPool = nn.AdaptiveAvgPool2d((None,1))

        self.SequenceModeling = nn.Sequential(

            BidirectionalLSTM(
                512,
                HIDDEN_SIZE,
                HIDDEN_SIZE
            ),

            BidirectionalLSTM(
                HIDDEN_SIZE,
                HIDDEN_SIZE,
                HIDDEN_SIZE
            )
        )

        self.Prediction = Attention(
            HIDDEN_SIZE,
            HIDDEN_SIZE,
            NUM_CLASSES
        )

    def forward(
        self,
        images,
        text,
        is_train=True
    ):

        visual_feature = self.FeatureExtraction(images)

        visual_feature = self.AdaptiveAvgPool(
            visual_feature.permute(0,3,1,2)
        )

        visual_feature = visual_feature.squeeze(3)

        contextual_feature = self.SequenceModeling(
            visual_feature
        )

        prediction = self.Prediction(
            contextual_feature.contiguous(),
            text,
            is_train,
            MAX_LABEL_LEN
        )

        return prediction

# ============================================================
# METRICS
# ============================================================

def compute_metrics(logits, labels):

    preds = logits.argmax(dim=-1)

    total_words = 0
    correct_words = 0

    total_chars = 0
    correct_chars = 0

    for pred_seq, gt_seq in zip(preds, labels):

        pred_str = decode_label(pred_seq)
        gt_str = decode_label(gt_seq)

        total_words += 1

        if pred_str == gt_str:
            correct_words += 1

        max_len = max(len(pred_str), len(gt_str))

        for i in range(max_len):

            p = pred_str[i] if i < len(pred_str) else ""
            g = gt_str[i] if i < len(gt_str) else ""

            total_chars += 1

            if p == g:
                correct_chars += 1

    word_acc = correct_words / total_words

    char_acc = correct_chars / max(total_chars, 1)

    return word_acc, char_acc

# ============================================================
# EVALUATION
# ============================================================

def evaluate(model, loader):

    model.eval()

    total_loss = 0

    total_wacc = 0
    total_cacc = 0

    n = 0

    with torch.no_grad():

        for images, labels, _ in loader:

            images = images.to(device)
            labels = labels.to(device)

            dummy_text = torch.zeros(
                labels.size(0),
                MAX_LABEL_LEN + 1,
                dtype=torch.long,
                device=device
            )

            logits = model(
                images,
                dummy_text,
                is_train=False
            )

            target = labels[:, 1:]

            loss = criterion(
                logits.reshape(-1, NUM_CLASSES),
                target.reshape(-1)
            )

            wacc, cacc = compute_metrics(
                logits.cpu(),
                labels.cpu()
            )

            total_loss += loss.item()
            total_wacc += wacc
            total_cacc += cacc

            n += 1

    return (
        total_loss / n,
        total_wacc / n,
        total_cacc / n
    )

# ============================================================
# LOAD DATA
# ============================================================

all_samples = load_iam_words(
    GT_FILE,
    DATASET_ROOT,
    MAX_LABEL_LEN
)

train_samples, val_samples, test_samples = make_splits(all_samples)

train_ds = IAMDataset(
    train_samples,
    IMG_H,
    IMG_W,
    MAX_LABEL_LEN,
    augment=True
)

val_ds = IAMDataset(
    val_samples,
    IMG_H,
    IMG_W,
    MAX_LABEL_LEN,
    augment=False
)

test_ds = IAMDataset(
    test_samples,
    IMG_H,
    IMG_W,
    MAX_LABEL_LEN,
    augment=False
)

train_loader = DataLoader(
    train_ds,
    batch_size=BATCH_SIZE,
    shuffle=True,
    num_workers=NUM_WORKERS,
    pin_memory=True
)

val_loader = DataLoader(
    val_ds,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=NUM_WORKERS,
    pin_memory=True
)

test_loader = DataLoader(
    test_ds,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=NUM_WORKERS,
    pin_memory=True
)

# ============================================================
# MODEL
# ============================================================

model = OCRModel().to(device)

print()
print("Trainable Parameters =", sum(
    p.numel()
    for p in model.parameters()
    if p.requires_grad
))

# ============================================================
# LOSS + OPTIMIZER
# ============================================================

criterion = nn.CrossEntropyLoss(
    ignore_index=PAD_IDX,
    label_smoothing=0.1
)

optimizer = optim.Adam(
    model.parameters(),
    lr=LEARNING_RATE
)

scheduler = optim.lr_scheduler.CosineAnnealingLR(
    optimizer,
    T_max=NUM_EPOCHS
)

# ============================================================
# TRAINING
# ============================================================

best_val_acc = 0

history = {
    "train_loss": [],
    "val_loss": [],
    "val_wacc": [],
    "val_cacc": []
}

epochs_no_improve = 0

print("\nStarting Training...\n")

for epoch in range(NUM_EPOCHS):

    model.train()

    running_loss = 0

    pbar = tqdm(train_loader)

    for images, labels, _ in pbar:

        images = images.to(device)
        labels = labels.to(device)

        logits = model(
            images,
            labels[:, :-1],
            is_train=True
        )

        target = labels[:, 1:]

        loss = criterion(
            logits.reshape(-1, NUM_CLASSES),
            target.reshape(-1)
        )

        optimizer.zero_grad()

        loss.backward()

        torch.nn.utils.clip_grad_norm_(
            model.parameters(),
            5.0
        )

        optimizer.step()

        running_loss += loss.item()

        pbar.set_description(
            f"Epoch {epoch+1}/{NUM_EPOCHS}"
        )

        pbar.set_postfix(
            loss=f"{loss.item():.4f}"
        )

    scheduler.step()

    train_loss = running_loss / len(train_loader)

    val_loss, val_wacc, val_cacc = evaluate(
        model,
        val_loader
    )

    history["train_loss"].append(train_loss)
    history["val_loss"].append(val_loss)
    history["val_wacc"].append(val_wacc)
    history["val_cacc"].append(val_cacc)

    print()
    print("="*50)
    print(f"Epoch {epoch+1}")
    print("="*50)
    print(f"Train Loss : {train_loss:.4f}")
    print(f"Val Loss   : {val_loss:.4f}")
    print(f"Val WAcc   : {val_wacc*100:.2f}%")
    print(f"Val CAcc   : {val_cacc*100:.2f}%")

    torch.save(
        model.state_dict(),
        os.path.join(SAVE_DIR, "last.pth")
    )

    if val_wacc > best_val_acc:

        best_val_acc = val_wacc

        torch.save(
            model.state_dict(),
            os.path.join(SAVE_DIR, "best.pth")
        )

        print("BEST MODEL SAVED")

        epochs_no_improve = 0

    else:
        epochs_no_improve += 1

    if epochs_no_improve >= PATIENCE:

        print("\nEarly stopping triggered.")
        break

# ============================================================
# FINAL TEST EVALUATION
# ============================================================

print("\nLoading Best Model...\n")

model.load_state_dict(
    torch.load(
        os.path.join(SAVE_DIR, "best.pth")
    )
)

test_loss, test_wacc, test_cacc = evaluate(
    model,
    test_loader
)

print("="*60)
print("FINAL TEST RESULTS")
print("="*60)

print(f"Test Loss : {test_loss:.4f}")
print(f"Test WAcc : {test_wacc*100:.2f}%")
print(f"Test CAcc : {test_cacc*100:.2f}%")

# ============================================================
# PLOT
# ============================================================

plt.figure(figsize=(12,5))

plt.plot(history["train_loss"], label="Train Loss")
plt.plot(history["val_loss"], label="Val Loss")

plt.legend()
plt.grid(True)

plt.savefig(
    os.path.join(SAVE_DIR, "loss_curve.png")
)

plt.show()

# ============================================================
# VISUALIZE PREDICTIONS
# ============================================================

print("\nShowing Predictions...\n")

model.eval()

images, labels, gt_texts = next(iter(test_loader))

images = images.to(device)

with torch.no_grad():

    dummy_text = torch.zeros(
        images.size(0),
        MAX_LABEL_LEN + 1,
        dtype=torch.long,
        device=device
    )

    logits = model(
        images,
        dummy_text,
        is_train=False
    )

preds = logits.argmax(dim=-1)

fig, axes = plt.subplots(3, 4, figsize=(16,10))

for i, ax in enumerate(axes.flat):

    if i >= len(images):
        break

    img = images[i].cpu().squeeze(0).numpy()

    img = (img * 0.5) + 0.5

    pred_text = decode_label(preds[i].cpu())

    gt_text = gt_texts[i]

    correct = pred_text == gt_text

    ax.imshow(img, cmap='gray')

    ax.set_title(
        f"GT: {gt_text}\nP: {pred_text}",
        color='green' if correct else 'red',
        fontsize=10
    )

    ax.axis('off')

plt.tight_layout()

plt.savefig(
    os.path.join(SAVE_DIR, "predictions.png")
)

plt.show()

print("\nTraining Complete")
```

```python
# ============================================================
# IAM WORD OCR — TEST SCRIPT
# Compatible with your TRAINING CODE
# Computes:
# 1. Word Accuracy
# 2. Character Accuracy
# 3. CER (Character Error Rate)
# 4. WER (Word Error Rate)
# ============================================================

import os
import random
import numpy as np
from PIL import Image

import torch
import torch.nn as nn
import torch.nn.functional as F

from torch.utils.data import Dataset, DataLoader
import torchvision.transforms as transforms

from tqdm import tqdm

# ============================================================
# DEVICE
# ============================================================

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print(f"Device : {device}")

if torch.cuda.is_available():
    print(f"GPU : {torch.cuda.get_device_name(0)}")

# ============================================================
# CONFIG
# ============================================================

DATASET_ROOT = "/home/mca/Shweta/handwritten text detection/dataset/iam/iam_words"

GT_FILE = os.path.join(DATASET_ROOT, "words.txt")

MODEL_PATH = "./saved_iam_2/best.pth"

BATCH_SIZE = 32
NUM_WORKERS = 4

IMG_H = 32
IMG_W = 160

MAX_LABEL_LEN = 25

HIDDEN_SIZE = 256

TEST_SPLIT = 0.1

SEED = 42

# ============================================================
# FIX SEED
# ============================================================

random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)

if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

# ============================================================
# VOCAB
# ============================================================

CHARS = ' !"#$%&\'()*+,-./0123456789:;<=>?@ABCDEFGHIJKLMNOPQRSTUVWXYZ[\\]^_`abcdefghijklmnopqrstuvwxyz{|}~'

GO_IDX  = 0
EOS_IDX = 1
PAD_IDX = 2
UNK_IDX = 3

CHAR_TO_IDX = {c: i + 4 for i, c in enumerate(CHARS)}

IDX_TO_CHAR = {v: k for k, v in CHAR_TO_IDX.items()}

IDX_TO_CHAR.update({
    GO_IDX: "<GO>",
    EOS_IDX: "<EOS>",
    PAD_IDX: "<PAD>",
    UNK_IDX: "<UNK>"
})

NUM_CLASSES = len(CHARS) + 4

print("NUM_CLASSES =", NUM_CLASSES)

# ============================================================
# LABEL ENCODING / DECODING
# ============================================================

def encode_label(text, max_len):

    tokens = [GO_IDX]

    for c in text[:max_len]:
        tokens.append(CHAR_TO_IDX.get(c, UNK_IDX))

    tokens.append(EOS_IDX)

    while len(tokens) < max_len + 2:
        tokens.append(PAD_IDX)

    return torch.tensor(tokens[:max_len+2], dtype=torch.long)


def decode_label(tensor):

    chars = []

    for idx in tensor.tolist():

        if idx == EOS_IDX:
            break

        if idx in [GO_IDX, PAD_IDX, UNK_IDX]:
            continue

        chars.append(IDX_TO_CHAR.get(idx, ""))

    return "".join(chars)

# ============================================================
# TRANSFORMS
# ============================================================

transform = transforms.Compose([
    transforms.Grayscale(1),
    transforms.Resize((IMG_H, IMG_W)),
    transforms.ToTensor(),
    transforms.Normalize([0.5], [0.5])
])

# ============================================================
# LOAD IAM WORDS
# ============================================================

def load_iam_words(gt_file, dataset_root, max_label_len):

    samples = []

    skipped = 0

    with open(gt_file, "r", encoding="utf-8") as f:

        for line in f:

            line = line.strip()

            if line.startswith("#"):
                continue

            if len(line) == 0:
                continue

            parts = line.split()

            if len(parts) < 9:
                skipped += 1
                continue

            word_id = parts[0]

            status = parts[1]

            if status != "ok":
                skipped += 1
                continue

            label = " ".join(parts[8:])

            if len(label) == 0:
                skipped += 1
                continue

            if len(label) > max_label_len:
                skipped += 1
                continue

            # ------------------------------------------------
            # IAM PATH
            # ------------------------------------------------

            folder1 = word_id.split("-")[0]

            folder2 = "-".join(word_id.split("-")[:2])

            img_path = os.path.join(
                dataset_root,
                "words",
                folder1,
                folder2,
                word_id + ".png"
            )

            if not os.path.exists(img_path):
                skipped += 1
                continue

            samples.append((img_path, label))

    print(f"Loaded Samples : {len(samples)}")
    print(f"Skipped        : {skipped}")

    return samples

# ============================================================
# DATASET
# ============================================================

class IAMDataset(Dataset):

    def __init__(
        self,
        samples,
        imgH=32,
        imgW=160,
        max_label_len=25
    ):

        self.samples = samples

        self.transform = transform

        self.max_label_len = max_label_len

    def __len__(self):

        return len(self.samples)

    def __getitem__(self, idx):

        img_path, label = self.samples[idx]

        try:
            image = Image.open(img_path).convert("RGB")

        except:
            image = Image.new("RGB", (IMG_W, IMG_H), (255,255,255))

        image = self.transform(image)

        encoded = encode_label(
            label,
            self.max_label_len
        )

        return image, encoded, label

# ============================================================
# TRAIN / VAL / TEST SPLIT
# ============================================================

def make_splits(samples):

    random.shuffle(samples)

    total = len(samples)

    n_train = int(total * 0.8)
    n_val   = int(total * 0.1)

    train_samples = samples[:n_train]

    val_samples = samples[n_train:n_train+n_val]

    test_samples = samples[n_train+n_val:]

    print()
    print("Dataset Split")
    print("-----------------------")
    print("Train :", len(train_samples))
    print("Val   :", len(val_samples))
    print("Test  :", len(test_samples))

    return train_samples, val_samples, test_samples

# ============================================================
# CNN FEATURE EXTRACTOR
# ============================================================

class CNNFeatureExtractor(nn.Module):

    def __init__(self):

        super().__init__()

        self.cnn = nn.Sequential(

            nn.Conv2d(1, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2,2),

            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2,2),

            nn.Conv2d(128, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),

            nn.Conv2d(256, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),

            nn.MaxPool2d((2,1), (2,1)),

            nn.Conv2d(256, 512, 3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True),

            nn.Conv2d(512, 512, 3, padding=1),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True),

            nn.MaxPool2d((2,1), (2,1)),

            nn.Conv2d(512, 512, 2),
            nn.BatchNorm2d(512),
            nn.ReLU(inplace=True)
        )

    def forward(self, x):

        return self.cnn(x)

# ============================================================
# BiLSTM
# ============================================================

class BidirectionalLSTM(nn.Module):

    def __init__(
        self,
        input_size,
        hidden_size,
        output_size
    ):

        super().__init__()

        self.rnn = nn.LSTM(
            input_size,
            hidden_size,
            bidirectional=True,
            batch_first=True
        )

        self.linear = nn.Linear(
            hidden_size * 2,
            output_size
        )

    def forward(self, x):

        x, _ = self.rnn(x)

        x = self.linear(x)

        return x

# ============================================================
# ATTENTION CELL
# ============================================================

class AttentionCell(nn.Module):

    def __init__(
        self,
        input_size,
        hidden_size,
        num_embeddings
    ):

        super().__init__()

        self.i2h = nn.Linear(
            input_size,
            hidden_size,
            bias=False
        )

        self.h2h = nn.Linear(
            hidden_size,
            hidden_size
        )

        self.score = nn.Linear(
            hidden_size,
            1,
            bias=False
        )

        self.rnn = nn.LSTMCell(
            input_size + num_embeddings,
            hidden_size
        )

    def forward(
        self,
        prev_hidden,
        batch_H,
        char_onehots
    ):

        batch_H_proj = self.i2h(batch_H)

        prev_hidden_proj = self.h2h(prev_hidden[0]).unsqueeze(1)

        e = self.score(
            torch.tanh(batch_H_proj + prev_hidden_proj)
        )

        alpha = F.softmax(e, dim=1)

        context = torch.bmm(
            alpha.permute(0,2,1),
            batch_H
        ).squeeze(1)

        concat_context = torch.cat(
            [context, char_onehots],
            dim=1
        )

        cur_hidden = self.rnn(
            concat_context,
            prev_hidden
        )

        return cur_hidden, alpha

# ============================================================
# ATTENTION
# ============================================================

class Attention(nn.Module):

    def __init__(
        self,
        input_size,
        hidden_size,
        num_classes
    ):

        super().__init__()

        self.attention_cell = AttentionCell(
            input_size,
            hidden_size,
            num_classes
        )

        self.hidden_size = hidden_size

        self.num_classes = num_classes

        self.generator = nn.Linear(
            hidden_size,
            num_classes
        )

    def _char_to_onehot(self, input_char):

        one_hot = torch.zeros(
            input_char.size(0),
            self.num_classes,
            device=input_char.device
        )

        one_hot.scatter_(1, input_char.unsqueeze(1), 1)

        return one_hot

    def forward(
        self,
        batch_H,
        text=None,
        is_train=False,
        batch_max_length=25
    ):

        batch_size = batch_H.size(0)

        num_steps = batch_max_length + 1

        hidden = (
            torch.zeros(batch_size, self.hidden_size, device=batch_H.device),
            torch.zeros(batch_size, self.hidden_size, device=batch_H.device)
        )

        targets = torch.LongTensor(
            batch_size
        ).fill_(GO_IDX).to(batch_H.device)

        probs = torch.zeros(
            batch_size,
            num_steps,
            self.num_classes,
            device=batch_H.device
        )

        for i in range(num_steps):

            char_onehots = self._char_to_onehot(targets)

            hidden, alpha = self.attention_cell(
                hidden,
                batch_H,
                char_onehots
            )

            probs_step = self.generator(hidden[0])

            probs[:, i, :] = probs_step

            _, next_input = probs_step.max(1)

            targets = next_input

        return probs

# ============================================================
# OCR MODEL
# ============================================================

class OCRModel(nn.Module):

    def __init__(self):

        super().__init__()

        self.FeatureExtraction = CNNFeatureExtractor()

        self.AdaptiveAvgPool = nn.AdaptiveAvgPool2d((None,1))

        self.SequenceModeling = nn.Sequential(

            BidirectionalLSTM(
                512,
                HIDDEN_SIZE,
                HIDDEN_SIZE
            ),

            BidirectionalLSTM(
                HIDDEN_SIZE,
                HIDDEN_SIZE,
                HIDDEN_SIZE
            )
        )

        self.Prediction = Attention(
            HIDDEN_SIZE,
            HIDDEN_SIZE,
            NUM_CLASSES
        )

    def forward(self, images):

        visual_feature = self.FeatureExtraction(images)

        visual_feature = self.AdaptiveAvgPool(
            visual_feature.permute(0,3,1,2)
        )

        visual_feature = visual_feature.squeeze(3)

        contextual_feature = self.SequenceModeling(
            visual_feature
        )

        prediction = self.Prediction(
            contextual_feature.contiguous(),
            None,
            is_train=False,
            batch_max_length=MAX_LABEL_LEN
        )

        return prediction

# ============================================================
# LEVENSHTEIN DISTANCE
# ============================================================

def levenshtein(s1, s2):

    m = len(s1)
    n = len(s2)

    dp = [[0]*(n+1) for _ in range(m+1)]

    for i in range(m+1):
        dp[i][0] = i

    for j in range(n+1):
        dp[0][j] = j

    for i in range(1, m+1):

        for j in range(1, n+1):

            cost = 0 if s1[i-1] == s2[j-1] else 1

            dp[i][j] = min(
                dp[i-1][j] + 1,
                dp[i][j-1] + 1,
                dp[i-1][j-1] + cost
            )

    return dp[m][n]

# ============================================================
# LOAD DATA
# ============================================================

all_samples = load_iam_words(
    GT_FILE,
    DATASET_ROOT,
    MAX_LABEL_LEN
)

_, _, test_samples = make_splits(all_samples)

test_ds = IAMDataset(
    test_samples,
    IMG_H,
    IMG_W,
    MAX_LABEL_LEN
)

test_loader = DataLoader(
    test_ds,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=NUM_WORKERS,
    pin_memory=True
)

# ============================================================
# LOAD MODEL
# ============================================================

model = OCRModel().to(device)

model.load_state_dict(
    torch.load(MODEL_PATH, map_location=device)
)

model.eval()

print("\nModel Loaded Successfully")

# ============================================================
# TESTING
# ============================================================

total_words = 0
correct_words = 0

total_chars = 0
correct_chars = 0

total_cer_dist = 0
total_cer_chars = 0

total_wer_dist = 0
total_wer_words = 0

print("\nRunning Evaluation...\n")

with torch.no_grad():

    for images, labels, gt_texts in tqdm(test_loader):

        images = images.to(device)

        logits = model(images)

        preds = logits.argmax(dim=-1)

        for pred_seq, gt_text in zip(preds, gt_texts):

            pred_text = decode_label(pred_seq.cpu())

            # ============================================
            # WORD ACCURACY
            # ============================================

            total_words += 1

            if pred_text == gt_text:
                correct_words += 1

            # ============================================
            # CHARACTER ACCURACY
            # ============================================

            max_len = max(len(pred_text), len(gt_text))

            for i in range(max_len):

                p = pred_text[i] if i < len(pred_text) else ""
                g = gt_text[i] if i < len(gt_text) else ""

                total_chars += 1

                if p == g:
                    correct_chars += 1

            # ============================================
            # CER
            # ============================================

            cer_dist = levenshtein(pred_text, gt_text)

            total_cer_dist += cer_dist

            total_cer_chars += len(gt_text)

            # ============================================
            # WER
            # ============================================

            pred_words = pred_text.split()
            gt_words = gt_text.split()

            wer_dist = levenshtein(pred_words, gt_words)

            total_wer_dist += wer_dist

            total_wer_words += len(gt_words)

# ============================================================
# FINAL METRICS
# ============================================================

word_acc = 100.0 * correct_words / total_words

char_acc = 100.0 * correct_chars / max(total_chars, 1)

cer = 100.0 * total_cer_dist / max(total_cer_chars, 1)

wer = 100.0 * total_wer_dist / max(total_wer_words, 1)

# ============================================================
# RESULTS
# ============================================================

print("\n======================================================")
print("IAM TEST RESULTS")
print("======================================================")

print(f"Total Samples        : {total_words}")
print(f"Correct Words        : {correct_words}")

print("\n---------------- METRICS ----------------")

print(f"Word Accuracy        : {word_acc:.2f}%")
print(f"Character Accuracy   : {char_acc:.2f}%")

print(f"\nCER                  : {cer:.2f}%")
print(f"WER                  : {wer:.2f}%")

print("======================================================")
```

```python
# ============================================================
# HVLT FOR IAM HANDWRITTEN WORD RECOGNITION
# FULL SINGLE-CELL TRAINING AND EVALUATION NOTEBOOK
# ============================================================

# ============================================================
# INSTALLS (RUN ONCE IF NEEDED)
# ============================================================
# !pip install timm transformers jiwer opencv-python pillow tqdm scikit-learn

# ============================================================
# IMPORTS
# ============================================================
import os
import cv2
import time
import random
import unicodedata
import numpy as np
from PIL import Image
from tqdm import tqdm
from sklearn.model_selection import train_test_split

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.optim import Adam
from torch.amp import autocast, GradScaler

import timm
from transformers import XLMRobertaModel
from jiwer import wer

# ============================================================
# CONFIG
# ============================================================
ROOT_DIR = "/home/mca/Shweta/handwritten text detection/dataset/iam/iam_words"
ORIGINAL_GT = os.path.join(ROOT_DIR, "words.txt")

IMG_HEIGHT = 32
IMG_WIDTH  = 128  # Optimally sized for standard individual IAM word slices

BATCH_SIZE = 32
LR = 5e-5
MAX_EPOCHS = 20
MAX_SEQ_LEN = 30
PATIENCE = 5

DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
NUM_WORKERS = 4
SEED = 42
USE_AMP = True

# ============================================================
# SEED
# ============================================================
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

set_seed(SEED)

# ============================================================
# IAM ENGLISH VOCABULARY
# ============================================================
IAM_CHARS = (
    "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
    "0123456789"
    ".,!?()[]{}:;-_'/\\&%$#@+*=\"„“ "
)

SPECIAL_TOKENS = ["<PAD>", "<SOS>", "<EOS>", "<UNK>"]
ALL_TOKENS = SPECIAL_TOKENS + list(IAM_CHARS)

char2idx = {c: i for i, c in enumerate(ALL_TOKENS)}
idx2char = {i: c for c, i in char2idx.items()}

VOCAB_SIZE = len(ALL_TOKENS)

PAD_IDX = char2idx["<PAD>"]
SOS_IDX = char2idx["<SOS>"]
EOS_IDX = char2idx["<EOS>"]
UNK_IDX = char2idx["<UNK>"]

print("VOCAB SIZE:", VOCAB_SIZE)

# ============================================================
# ROBUST DATA EXTRACTION ENGINE WITH FILE INTEGRITY CHECK
# ============================================================
samples = []
missing_count = 0
corrupted_count = 0

print("Parsing and validating IAM metadata registry structure...")
with open(ORIGINAL_GT, "r", encoding="utf-8", errors="ignore") as f:
    for line in f:
        line = line.strip()
        if not line or line.startswith("#"):
            continue
            
        parts = line.split()
        if len(parts) < 9 or parts[1] == "err":
            continue
            
        word_id = parts[0]
        text_transcription = " ".join(parts[8:])
        
        # Build strict relative directory structure mapping
        id_splits = word_id.split("-")
        f1 = id_splits[0]
        f2 = f"{id_splits[0]}-{id_splits[1]}"
        rel_img_path = f"words/{f1}/{f2}/{word_id}.png"
        full_img_path = os.path.join(ROOT_DIR, rel_img_path)
        
        # Core physical disk confirmation check
        if not os.path.exists(full_img_path):
            missing_count += 1
            continue
            
        try:
            with Image.open(full_img_path) as img:
                img.verify()
            text_transcription = unicodedata.normalize("NFC", text_transcription.strip())
            samples.append((full_img_path, text_transcription))
        except Exception:
            corrupted_count += 1
            continue

print(f"TOTAL SAMPLES VALIDATED : {len(samples)}")
print(f"SKIPPED MISSING IMAGES  : {missing_count}")
print(f"SKIPPED CORRUPTED IMAGES: {corrupted_count}")

if len(samples) == 0:
    raise ValueError("Execution halted. Verification discovered zero processing items.")

# ============================================================
# STRATIFIED EXPERIMENTAL DATA SPLITTING
# ============================================================
train_samples, temp_samples = train_test_split(
    samples, test_size=0.15, random_state=SEED
)
val_samples, test_samples = train_test_split(
    temp_samples, test_size=0.5, random_state=SEED
)

print(f"TRAIN SAMPLES : {len(train_samples)}")
print(f"VAL SAMPLES   : {len(val_samples)}")
print(f"TEST SAMPLES  : {len(test_samples)}")

# ============================================================
# IMAGE PREPROCESSING ENGINE WITH DATA AUGMENTATION
# ============================================================
def preprocess_image(img_path, augment=False):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    if augment:
        # Subtle Brightness adjustments to combat exposure variations
        if random.random() < 0.5:
            alpha = random.uniform(0.8, 1.2)
            beta = random.randint(-15, 15)
            img = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
        # Random Blur transformations to improve stroke feature extraction
        if random.random() < 0.3:
            k = random.choice([3, 5])
            img = cv2.GaussianBlur(img, (k, k), 0)

    h, w = img.shape[:2]
    scale = IMG_HEIGHT / h
    new_w = max(1, int(w * scale))

    img = cv2.resize(img, (new_w, IMG_HEIGHT))

    if new_w < IMG_WIDTH:
        pad = np.ones((IMG_HEIGHT, IMG_WIDTH - new_w, 3), dtype=np.uint8) * 255
        img = np.concatenate([img, pad], axis=1)
    else:
        img = cv2.resize(img, (IMG_WIDTH, IMG_HEIGHT))

    img = img.astype(np.float32) / 255.0
    img = (img - 0.5) / 0.5
    img = np.transpose(img, (2, 0, 1))

    return torch.tensor(img, dtype=torch.float32)

# ============================================================
# SEQUENTIAL TOKEN TRANSFORMATIONS
# ============================================================
def encode_text(text):
    tokens = [SOS_IDX]
    for ch in text:
        tokens.append(char2idx.get(ch, UNK_IDX))
    tokens.append(EOS_IDX)

    if len(tokens) < MAX_SEQ_LEN:
        tokens += [PAD_IDX] * (MAX_SEQ_LEN - len(tokens))
    else:
        tokens = tokens[:MAX_SEQ_LEN]
        tokens[-1] = EOS_IDX

    return torch.tensor(tokens, dtype=torch.long)

def decode_tokens(tokens):
    chars = []
    for t in tokens:
        t = int(t)
        if t == EOS_IDX:
            break
        if t in [PAD_IDX, SOS_IDX]:
            continue
        chars.append(idx2char.get(t, ""))
    return "".join(chars)

# ============================================================
# TORCH COHESIVE DATA LAYERS
# ============================================================
class IAMWordDataset(Dataset):
    def __init__(self, dataset_samples, augment=False):
        self.samples = dataset_samples
        self.augment = augment

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        img_path, text = self.samples[idx]
        image = preprocess_image(img_path, augment=self.augment)
        tokens = encode_text(text)
        return image, tokens, text

train_loader = DataLoader(
    IAMWordDataset(train_samples, augment=True),
    batch_size=BATCH_SIZE, shuffle=True, num_workers=NUM_WORKERS, pin_memory=True
)
val_loader = DataLoader(
    IAMWordDataset(val_samples, augment=False),
    batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True
)
test_loader = DataLoader(
    IAMWordDataset(test_samples, augment=False),
    batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True
)

# ============================================================
# POSITIONAL BRIDGE
# ============================================================
class PositionalBridge(nn.Module):
    def __init__(self, in_channels=1024, d_model=768, vis_seq_len=256):
        super().__init__()
        self.pool = nn.AdaptiveAvgPool2d((1, vis_seq_len))
        self.proj = nn.Linear(in_channels, d_model)
        self.pos_embed = nn.Parameter(torch.randn(1, vis_seq_len, d_model) * 0.02)

    def forward(self, x):
        x = x.permute(0, 3, 1, 2)
        x = self.pool(x).squeeze(2).permute(0, 2, 1)
        return self.proj(x) + self.pos_embed

# ============================================================
# TRANSFORMER DECODER
# ============================================================
class IAMDecoder(nn.Module):
    def __init__(self, vocab_size, d_model=768, n_heads=12, n_layers=6):
        super().__init__()
        self.token_embed = nn.Embedding(vocab_size, d_model, padding_idx=PAD_IDX)
        self.pos_embed = nn.Embedding(MAX_SEQ_LEN, d_model)

        # In IAMDecoder.__init__, change:
        decoder_layer = nn.TransformerDecoderLayer(
            d_model=d_model, nhead=n_heads, dim_feedforward=3072,
            dropout=0.3,   # was 0.1 — increase significantly
            batch_first=True
        )
        
        
        self.decoder = nn.TransformerDecoder(decoder_layer, num_layers=n_layers)
        self.output_proj = nn.Linear(d_model, vocab_size)

        print("Aligning Pretrained Multilingual Token Embedding Weights...")
        try:
            xlm = XLMRobertaModel.from_pretrained("xlm-roberta-base")
            emb = xlm.embeddings.word_embeddings.weight
            n = min(emb.shape[0], vocab_size)
            self.token_embed.weight.data[:n] = emb[:n]
            # Also add dropout after token embedding:
            self.embed_dropout = nn.Dropout(0.2)
            del xlm
            print("Successfully embedded weight alignment layers.")
        except Exception as e:
            print("Pretrained initializing step bypassed:", e)

    def forward(self, memory, tgt_tokens):
        B, T = tgt_tokens.shape
        pos = torch.arange(T, device=tgt_tokens.device).unsqueeze(0).expand(B, -1)
        x = self.embed_dropout(self.token_embed(tgt_tokens) + self.pos_embed(pos))

        tgt_mask = torch.triu(torch.ones(T, T, device=tgt_tokens.device), diagonal=1).bool()
        tgt_key_padding_mask = (tgt_tokens == PAD_IDX)

        out = self.decoder(
            tgt=x, memory=memory, tgt_mask=tgt_mask, tgt_key_padding_mask=tgt_key_padding_mask
        )
        return self.output_proj(out)

    @torch.no_grad()
    def greedy_decode(self, memory, max_len=MAX_SEQ_LEN):
        B = memory.shape[0]
        generated = torch.full((B, 1), SOS_IDX, device=memory.device, dtype=torch.long)
        finished = torch.zeros(B, dtype=torch.bool, device=memory.device)

        for _ in range(max_len):
            logits = self.forward(memory, generated)
            next_token = logits[:, -1].argmax(dim=-1)
            next_token = next_token.masked_fill(finished, PAD_IDX)

            generated = torch.cat([generated, next_token.unsqueeze(1)], dim=1)
            finished |= (next_token == EOS_IDX)

            if finished.all():
                break
        return generated[:, 1:]

# ============================================================
# COMBINED HVLT NET ENGINE 
# ============================================================
class HVLT(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = timm.create_model(
            "swin_base_patch4_window7_224",
            pretrained=True, features_only=True,
            img_size=(32, 192), strict_img_size=False,
        )
        self.bridge = PositionalBridge(in_channels=1024, d_model=768, vis_seq_len=256)
        self.decoder = IAMDecoder(vocab_size=VOCAB_SIZE)

    def forward(self, images, tgt_tokens):
        feats = self.encoder(images)[-1]
        memory = self.bridge(feats)
        return self.decoder(memory, tgt_tokens)

    @torch.no_grad()
    def predict(self, images):
        feats = self.encoder(images)[-1]
        memory = self.bridge(feats)
        return self.decoder.greedy_decode(memory)

# ============================================================
# EVALUATION METRICS
# ============================================================
criterion = nn.CrossEntropyLoss(ignore_index=PAD_IDX)

def char_accuracy(preds, labels):
    correct, total = 0, 0
    for p, l in zip(preds, labels):
        n = min(len(p), len(l))
        for i in range(n):
            if p[i] == l[i]:
                correct += 1
        total += max(len(p), len(l))
    return 100.0 * correct / max(total, 1)

# ============================================================
# SYSTEM MODEL INSTANTIATION
# ============================================================
model = HVLT().to(DEVICE)
optimizer = Adam(model.parameters(), lr=LR)
scaler = GradScaler("cuda", enabled=USE_AMP)

print("TOTAL EXECUTABLE PARAMETERS:", sum(p.numel() for p in model.parameters()) / 1e6, "M")

# ============================================================
# MAIN ACTIVE TRAINING PIPELINE LOOP
# ============================================================
best_val_loss = float('inf')
patience_counter = 0

for epoch in range(1, MAX_EPOCHS + 1):
    # --- TRAINING STAGE ---
    model.train()
    train_loss = 0
    train_preds, train_labels = [], []
    
    pbar = tqdm(train_loader)
    for images, targets, labels in pbar:
        images, targets = images.to(DEVICE), targets.to(DEVICE)
        decoder_input = targets[:, :-1]
        decoder_target = targets[:, 1:]

        optimizer.zero_grad()
        with autocast("cuda", enabled=USE_AMP):
            logits = model(images, decoder_input)
            loss = criterion(logits.reshape(-1, VOCAB_SIZE), decoder_target.reshape(-1))

        scaler.scale(loss).backward()
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(model.parameters(), 5.0)
        scaler.step(optimizer)
        scaler.update()

        train_loss += loss.item()

        with torch.no_grad():
            pred_ids = logits.argmax(dim=-1)
            preds = [decode_tokens(x) for x in pred_ids]

        train_preds.extend(preds)
        train_labels.extend(labels)
        pbar.set_description(f"Epoch {epoch} | Train Loss: {loss.item():.4f}")

    train_cer = char_accuracy(train_preds, train_labels)
    train_wer = wer(train_labels, train_preds) * 100
    avg_train_loss = train_loss / len(train_loader)

    # --- VALIDATION STAGE ---
    model.eval()
    val_loss = 0
    val_preds, val_labels = [], []

    with torch.no_grad():
        for images, targets, labels in tqdm(val_loader, desc="Validating Loop"):
            images, targets = images.to(DEVICE), targets.to(DEVICE)
            decoder_input = targets[:, :-1]
            decoder_target = targets[:, 1:]

            logits = model(images, decoder_input)
            loss = criterion(logits.reshape(-1, VOCAB_SIZE), decoder_target.reshape(-1))
            val_loss += loss.item()

            pred_ids = model.predict(images)
            preds = [decode_tokens(x) for x in pred_ids]

            val_preds.extend(preds)
            val_labels.extend(labels)

    val_cer = char_accuracy(val_preds, val_labels)
    val_wer = wer(val_labels, val_preds) * 100
    avg_val_loss = val_loss / len(val_loader)

    # --- METRIC OUTPUT LOGGING ---
    print("\n" + "="*60)
    print(f"EPOCH {epoch}")
    print(f"TRAIN LOSS: {avg_train_loss:.4f} | VAL LOSS: {avg_val_loss:.4f}")
    print(f"TRAIN CAR:  {train_cer:.2f}%  | VAL CAR:  {val_cer:.2f}%")
    print(f"TRAIN WER:  {train_wer:.2f}%  | VAL WER:  {val_wer:.2f}%")
    print("="*60)

    # --- STABLE EARLY STOPPING LOGIC BASED ON LOSS OPTIMIZATION ---
    if avg_val_loss < best_val_loss:
        best_val_loss = avg_val_loss
        patience_counter = 0
        torch.save({
            "model": model.state_dict(),
            "epoch": epoch,
            "val_loss": avg_val_loss,
        }, "best_iam_hvlt.pt")
        print(">>> PERFORMANCE INCREASE DETECTED: BEST MODEL SAVED.")
    else:
        patience_counter += 1
        print(f">>> No improvement in Validation Loss. Early Stopping Counter: {patience_counter}/{PATIENCE}")

    if patience_counter >= PATIENCE:
        print(f"\n[EARLY STOPPING TRIGGERED] Validation stalled for {PATIENCE} epochs. Halting training.")
        break

# ============================================================
# FINAL INDEPENDENT TEST EVALUATION
# ============================================================
print("\nFINAL INDEPENDENT TEST EVALUATION RUNNING...")
if os.path.exists("best_iam_hvlt.pt"):
    checkpoint = torch.load("best_iam_hvlt.pt")
    model.load_state_dict(checkpoint["model"])

model.eval()
test_preds, test_labels = [], []

with torch.no_grad():
    for images, _, labels in tqdm(test_loader, desc="Testing Model"):
        images = images.to(DEVICE)
        pred_ids = model.predict(images)
        preds = [decode_tokens(x) for x in pred_ids]
        test_preds.extend(preds)
        test_labels.extend(labels)

test_cer = char_accuracy(test_preds, test_labels)
test_wer = wer(test_labels, test_preds) * 100

print("\n" + "#"*40 + "\nFINAL GENERALIZATION PERFORMANCE TEST METRICS")
print(f"TEST CAR (Character Accuracy Rate) : {test_cer:.2f}%")
print(f"TEST WER (Word Error Rate)       : {test_wer:.2f}%")
print("#"*40)

# ============================================================
# RANDOM EVALUATION PREVIEWS VISUAL VERIFICATION
# ============================================================
print("\nSAMPLE VISUALIZATION VERIFICATIONS:\n")
for i in range(min(10, len(test_labels))):
    print(f"GROUND TRUTH : {test_labels[i]}")
    print(f"PREDICTED    : {test_preds[i]}")
    print("-" * 50)
```

```python
# ============================================================
# HVLT FOR IAM HANDWRITTEN WORD RECOGNITION - IMPROVED VERSION
# Fixes from original:
#   1. IMG_WIDTH: 128 → 192, consistent with encoder img_size
#   2. encoder img_size: (32,192) properly matched to IMG dims
#   3. vis_seq_len: 256 → 192 (matches actual image width)
#   4. MAX_EPOCHS: 20 → 100, PATIENCE: 5 → 15
#   5. CosineAnnealingLR scheduler added
#   6. Stronger augmentation: rotation, elastic distortion
#   7. LLRD (layerwise learning rate decay) added
#   8. Beam search decoding at test time
#   9. Unicode fix for special chars in vocabulary
# ============================================================


# ============================================================
# IMPORTS
# ============================================================
import os
import cv2
import time
import random
import unicodedata
import numpy as np
from PIL import Image
from tqdm import tqdm
from sklearn.model_selection import train_test_split

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.optim import Adam
from torch.optim.lr_scheduler import CosineAnnealingLR
from torch.amp import autocast, GradScaler

import timm
from transformers import XLMRobertaModel
from jiwer import wer

# Optional stronger augmentation
# pip install albumentations
try:
    import albumentations as A
    HAS_ALBUMENTATIONS = True
except ImportError:
    HAS_ALBUMENTATIONS = False
    print("albumentations not installed. Using basic augmentation.")
    print("Install with: pip install albumentations")


# ============================================================
# CONFIG
# ============================================================
ROOT_DIR    = "/home/mca/Shweta/handwritten text detection/dataset/iam/iam_words"
ORIGINAL_GT = os.path.join(ROOT_DIR, "words.txt")

# FIX 1: IMG_WIDTH was 128 but encoder expected 192 — now consistent
IMG_HEIGHT = 32
IMG_WIDTH  = 192

BATCH_SIZE = 32
LR         = 5e-5

# FIX 2: was 20 epochs / patience 5 — far too short for IAM
MAX_EPOCHS = 100
PATIENCE   = 15

MAX_SEQ_LEN = 30
BEAM_WIDTH  = 5
LLRD_DECAY  = 0.75   # layerwise LR decay factor

DEVICE     = "cuda" if torch.cuda.is_available() else "cpu"
NUM_WORKERS = 4
SEED        = 42
USE_AMP     = True

CHECKPOINT_PATH = "best_iam_hvlt.pt"


# ============================================================
# SEED
# ============================================================
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

set_seed(SEED)


# ============================================================
# IAM VOCABULARY
# FIX 3: special chars written as unicode escapes to avoid
# SyntaxError with characters like „ " ¬
# ============================================================
IAM_CHARS = (
    "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
    "0123456789"
    ".,!?()[]{}:;-_'/\\&%$#@+*=\"\u201e\u201c "
)

SPECIAL_TOKENS = ["<PAD>", "<SOS>", "<EOS>", "<UNK>"]
ALL_TOKENS     = SPECIAL_TOKENS + list(IAM_CHARS)

char2idx  = {c: i for i, c in enumerate(ALL_TOKENS)}
idx2char  = {i: c for c, i in char2idx.items()}
VOCAB_SIZE = len(ALL_TOKENS)

PAD_IDX = char2idx["<PAD>"]
SOS_IDX = char2idx["<SOS>"]
EOS_IDX = char2idx["<EOS>"]
UNK_IDX = char2idx["<UNK>"]

print("VOCAB SIZE:", VOCAB_SIZE)


# ============================================================
# DATA LOADING
# ============================================================
samples        = []
missing_count  = 0
corrupted_count = 0

print("Parsing IAM metadata...")
with open(ORIGINAL_GT, "r", encoding="utf-8", errors="ignore") as f:
    for line in f:
        line = line.strip()
        if not line or line.startswith("#"):
            continue

        parts = line.split()
        if len(parts) < 9 or parts[1] == "err":
            continue

        word_id            = parts[0]
        text_transcription = " ".join(parts[8:])

        id_splits   = word_id.split("-")
        f1          = id_splits[0]
        f2          = f"{id_splits[0]}-{id_splits[1]}"
        rel_img_path = f"words/{f1}/{f2}/{word_id}.png"
        full_img_path = os.path.join(ROOT_DIR, rel_img_path)

        if not os.path.exists(full_img_path):
            missing_count += 1
            continue

        try:
            with Image.open(full_img_path) as img:
                img.verify()
            text_transcription = unicodedata.normalize("NFC", text_transcription.strip())
            samples.append((full_img_path, text_transcription))
        except Exception:
            corrupted_count += 1
            continue

print(f"TOTAL SAMPLES    : {len(samples)}")
print(f"MISSING IMAGES   : {missing_count}")
print(f"CORRUPTED IMAGES : {corrupted_count}")

if len(samples) == 0:
    raise ValueError("No samples found. Check ROOT_DIR path.")


# ============================================================
# DATA SPLITTING
# ============================================================
train_samples, temp_samples = train_test_split(samples, test_size=0.15, random_state=SEED)
val_samples, test_samples   = train_test_split(temp_samples, test_size=0.5, random_state=SEED)

print(f"TRAIN : {len(train_samples)}")
print(f"VAL   : {len(val_samples)}")
print(f"TEST  : {len(test_samples)}")


# ============================================================
# FIX 4: STRONGER AUGMENTATION
# IAM benefits from rotation (handwriting slant varies a lot)
# and elastic distortion (pen pressure, paper texture)
# ============================================================
if HAS_ALBUMENTATIONS:
    strong_aug = A.Compose([
        A.ElasticTransform(alpha=20, sigma=4, p=0.4),
        A.Rotate(limit=3, border_mode=cv2.BORDER_REPLICATE, p=0.4),
        A.GaussNoise(var_limit=(5, 20), p=0.3),
        A.RandomBrightnessContrast(brightness_limit=0.2, contrast_limit=0.2, p=0.5),
        A.GaussianBlur(blur_limit=(3, 5), p=0.2),
        A.Morphological(scale=(1, 2), operation='dilation', p=0.15),
        A.Morphological(scale=(1, 2), operation='erosion',  p=0.15),
    ])
else:
    strong_aug = None


def preprocess_image(img_path, augment=False):
    img = cv2.imread(img_path)
    if img is None:
        # Return blank image if file unreadable
        img = np.ones((IMG_HEIGHT, IMG_WIDTH, 3), dtype=np.uint8) * 255
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    if augment:
        if strong_aug is not None:
            img = strong_aug(image=img)["image"]
        else:
            # Fallback basic augmentation
            if random.random() < 0.5:
                alpha = random.uniform(0.8, 1.2)
                beta  = random.randint(-15, 15)
                img   = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
            if random.random() < 0.3:
                k   = random.choice([3, 5])
                img = cv2.GaussianBlur(img, (k, k), 0)
            # Rotation augmentation (even without albumentations)
            if random.random() < 0.4:
                angle = random.uniform(-3, 3)
                M     = cv2.getRotationMatrix2D(
                    (img.shape[1] // 2, img.shape[0] // 2), angle, 1.0
                )
                img = cv2.warpAffine(
                    img, M, (img.shape[1], img.shape[0]),
                    borderValue=(255, 255, 255)
                )

    h, w   = img.shape[:2]
    scale  = IMG_HEIGHT / h
    new_w  = max(1, int(w * scale))
    img    = cv2.resize(img, (new_w, IMG_HEIGHT))

    if new_w < IMG_WIDTH:
        pad = np.ones((IMG_HEIGHT, IMG_WIDTH - new_w, 3), dtype=np.uint8) * 255
        img = np.concatenate([img, pad], axis=1)
    else:
        img = cv2.resize(img, (IMG_WIDTH, IMG_HEIGHT))

    img = img.astype(np.float32) / 255.0
    img = (img - 0.5) / 0.5
    img = np.transpose(img, (2, 0, 1))

    return torch.tensor(img, dtype=torch.float32)


# ============================================================
# TEXT UTILS
# ============================================================
def encode_text(text):
    tokens = [SOS_IDX]
    for ch in text:
        tokens.append(char2idx.get(ch, UNK_IDX))
    tokens.append(EOS_IDX)

    if len(tokens) < MAX_SEQ_LEN:
        tokens += [PAD_IDX] * (MAX_SEQ_LEN - len(tokens))
    else:
        tokens    = tokens[:MAX_SEQ_LEN]
        tokens[-1] = EOS_IDX

    return torch.tensor(tokens, dtype=torch.long)


def decode_tokens(tokens):
    chars = []
    for t in tokens:
        t = int(t)
        if t == EOS_IDX:
            break
        if t in [PAD_IDX, SOS_IDX]:
            continue
        chars.append(idx2char.get(t, ""))
    return "".join(chars)


# ============================================================
# DATASET
# ============================================================
class IAMWordDataset(Dataset):
    def __init__(self, dataset_samples, augment=False):
        self.samples = dataset_samples
        self.augment = augment

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        img_path, text = self.samples[idx]
        image  = preprocess_image(img_path, augment=self.augment)
        tokens = encode_text(text)
        return image, tokens, text


train_loader = DataLoader(
    IAMWordDataset(train_samples, augment=True),
    batch_size=BATCH_SIZE, shuffle=True, num_workers=NUM_WORKERS, pin_memory=True
)
val_loader = DataLoader(
    IAMWordDataset(val_samples, augment=False),
    batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True
)
test_loader = DataLoader(
    IAMWordDataset(test_samples, augment=False),
    batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True
)


# ============================================================
# FIX 5: POSITIONAL BRIDGE — vis_seq_len matched to image width
# Original: vis_seq_len=256 but image was only 128px wide
# Now: vis_seq_len=192 to match IMG_WIDTH=192
# ============================================================
class PositionalBridge(nn.Module):
    def __init__(self, in_channels=1024, d_model=768, vis_seq_len=192):
        super().__init__()
        self.pool      = nn.AdaptiveAvgPool2d((1, vis_seq_len))
        self.proj      = nn.Linear(in_channels, d_model)
        self.pos_embed = nn.Parameter(torch.randn(1, vis_seq_len, d_model) * 0.02)

    def forward(self, x):
        x = x.permute(0, 3, 1, 2)
        x = self.pool(x).squeeze(2).permute(0, 2, 1)
        return self.proj(x) + self.pos_embed


# ============================================================
# DECODER WITH BEAM SEARCH
# ============================================================
class IAMDecoder(nn.Module):
    def __init__(self, vocab_size, d_model=768, n_heads=12, n_layers=6):
        super().__init__()
        self.token_embed = nn.Embedding(vocab_size, d_model, padding_idx=PAD_IDX)
        self.pos_embed   = nn.Embedding(MAX_SEQ_LEN, d_model)

        decoder_layer = nn.TransformerDecoderLayer(
            d_model=d_model, nhead=n_heads, dim_feedforward=3072,
            dropout=0.1, batch_first=True
        )
        self.decoder     = nn.TransformerDecoder(decoder_layer, num_layers=n_layers)
        self.output_proj = nn.Linear(d_model, vocab_size)

        print("Aligning Pretrained Token Embedding Weights...")
        try:
            xlm = XLMRobertaModel.from_pretrained("xlm-roberta-base")
            emb = xlm.embeddings.word_embeddings.weight
            n   = min(emb.shape[0], vocab_size)
            self.token_embed.weight.data[:n] = emb[:n]
            del xlm
            print("Weight alignment done.")
        except Exception as e:
            print("Pretrained init skipped:", e)

    def forward(self, memory, tgt_tokens):
        B, T = tgt_tokens.shape
        pos  = torch.arange(T, device=tgt_tokens.device).unsqueeze(0).expand(B, -1)
        x    = self.token_embed(tgt_tokens) + self.pos_embed(pos)

        tgt_mask             = torch.triu(torch.ones(T, T, device=tgt_tokens.device), diagonal=1).bool()
        tgt_key_padding_mask = (tgt_tokens == PAD_IDX)

        out = self.decoder(
            tgt=x, memory=memory,
            tgt_mask=tgt_mask,
            tgt_key_padding_mask=tgt_key_padding_mask
        )
        return self.output_proj(out)

    # FIX 6: Beam search — better WER than greedy at test time
    @torch.no_grad()
    def beam_decode(self, memory, beam_width=BEAM_WIDTH, max_len=MAX_SEQ_LEN):
        B       = memory.shape[0]
        results = []

        for b in range(B):
            mem       = memory[b:b+1]
            beams     = [(0.0, [SOS_IDX])]
            completed = []

            for _ in range(max_len):
                candidates = []
                for score, tokens in beams:
                    if tokens[-1] == EOS_IDX:
                        completed.append((score, tokens))
                        continue

                    t     = torch.tensor([tokens], dtype=torch.long, device=memory.device)
                    logit = self.forward(mem, t)
                    log_p = F.log_softmax(logit[0, -1], dim=-1)

                    topk_scores, topk_ids = log_p.topk(beam_width)
                    for s, idx in zip(topk_scores.tolist(), topk_ids.tolist()):
                        candidates.append((score + s, tokens + [idx]))

                if not candidates:
                    break

                candidates.sort(key=lambda x: x[0], reverse=True)
                beams = candidates[:beam_width]

                if all(t[-1] == EOS_IDX for _, t in beams):
                    completed.extend(beams)
                    break

            if not completed:
                completed = beams

            best_tokens = max(completed, key=lambda x: x[0])[1]
            results.append(torch.tensor(best_tokens[1:], dtype=torch.long))

        return results

    # Greedy kept for fast validation during training
    @torch.no_grad()
    def greedy_decode(self, memory, max_len=MAX_SEQ_LEN):
        B         = memory.shape[0]
        generated = torch.full((B, 1), SOS_IDX, device=memory.device, dtype=torch.long)
        finished  = torch.zeros(B, dtype=torch.bool, device=memory.device)

        for _ in range(max_len):
            logits     = self.forward(memory, generated)
            next_token = logits[:, -1].argmax(dim=-1)
            next_token = next_token.masked_fill(finished, PAD_IDX)

            generated = torch.cat([generated, next_token.unsqueeze(1)], dim=1)
            finished |= (next_token == EOS_IDX)

            if finished.all():
                break

        return generated[:, 1:]


# ============================================================
# MODEL
# FIX 7: img_size now matches IMG_HEIGHT x IMG_WIDTH exactly
# Original had img_size=(32,192) while IMG_WIDTH was 128 — mismatch
# ============================================================
class HVLT(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = timm.create_model(
            "swin_base_patch4_window7_224",
            pretrained=True, features_only=True,
            img_size=(IMG_HEIGHT, IMG_WIDTH), strict_img_size=False,
        )
        self.bridge  = PositionalBridge(in_channels=1024, d_model=768, vis_seq_len=192)
        self.decoder = IAMDecoder(vocab_size=VOCAB_SIZE)

    def forward(self, images, tgt_tokens):
        feats  = self.encoder(images)[-1]
        memory = self.bridge(feats)
        return self.decoder(memory, tgt_tokens)

    def _encode(self, images):
        feats  = self.encoder(images)[-1]
        return self.bridge(feats)

    @torch.no_grad()
    def predict(self, images, use_beam=True):
        memory = self._encode(images)
        if use_beam:
            return self.decoder.beam_decode(memory)
        return self.decoder.greedy_decode(memory)


# ============================================================
# LOSS & METRICS
# ============================================================
criterion = nn.CrossEntropyLoss(ignore_index=PAD_IDX)


def char_accuracy(preds, labels):
    correct, total = 0, 0
    for p, l in zip(preds, labels):
        n = min(len(p), len(l))
        for i in range(n):
            if p[i] == l[i]:
                correct += 1
        total += max(len(p), len(l))
    return 100.0 * correct / max(total, 1)


def decode_beam_results(beam_results):
    return [decode_tokens(t.tolist()) for t in beam_results]


# ============================================================
# FIX 8: LAYERWISE LEARNING RATE DECAY
# Earlier Swin stages (low-level features) get smaller LR
# Decoder/bridge get full LR — they need the most updating
# ============================================================
def get_llrd_param_groups(model, base_lr, decay):
    decoder_bridge_params = []
    stage4_params = []
    stage3_params = []
    stage2_params = []
    stage1_embed_params = []

    for name, param in model.named_parameters():
        if not param.requires_grad:
            continue

        if name.startswith("decoder.") or name.startswith("bridge."):
            decoder_bridge_params.append(param)
        elif "encoder.layers.3" in name or "encoder.norm3" in name:
            stage4_params.append(param)
        elif "encoder.layers.2" in name or "encoder.norm2" in name:
            stage3_params.append(param)
        elif "encoder.layers.1" in name or "encoder.norm1" in name:
            stage2_params.append(param)
        else:
            stage1_embed_params.append(param)

    param_groups = [
        {"params": decoder_bridge_params, "lr": base_lr},
        {"params": stage4_params,         "lr": base_lr * decay},
        {"params": stage3_params,         "lr": base_lr * decay**2},
        {"params": stage2_params,         "lr": base_lr * decay**3},
        {"params": stage1_embed_params,   "lr": base_lr * decay**4},
    ]

    group_names = ["decoder+bridge", "swin stage 4", "swin stage 3",
                   "swin stage 2",   "swin stage 1+embed"]
    print("\nLLRD param groups:")
    for n, g in zip(group_names, param_groups):
        print(f"  {n:25s}  lr={g['lr']:.2e}  ({len(g['params'])} tensors)")

    return param_groups


# ============================================================
# MODEL + OPTIMIZER + SCHEDULER
# ============================================================
model = HVLT().to(DEVICE)

print("\nBuilding LLRD optimizer:")
param_groups = get_llrd_param_groups(model, LR, LLRD_DECAY)
optimizer    = Adam(param_groups)

# FIX 9: Cosine LR scheduler — was missing entirely in original
scheduler = CosineAnnealingLR(optimizer, T_max=MAX_EPOCHS, eta_min=1e-7)

scaler = GradScaler("cuda", enabled=USE_AMP)

print("\nTOTAL PARAMS:", sum(p.numel() for p in model.parameters()) / 1e6, "M")


# ============================================================
# TRAINING LOOP
# ============================================================
best_val_loss    = float('inf')
patience_counter = 0

for epoch in range(1, MAX_EPOCHS + 1):

    # --- TRAIN ---
    model.train()
    train_loss = 0
    train_preds, train_labels = [], []

    pbar = tqdm(train_loader)
    for images, targets, labels in pbar:
        images, targets = images.to(DEVICE), targets.to(DEVICE)
        decoder_input  = targets[:, :-1]
        decoder_target = targets[:, 1:]

        optimizer.zero_grad()
        with autocast("cuda", enabled=USE_AMP):
            logits = model(images, decoder_input)
            loss   = criterion(logits.reshape(-1, VOCAB_SIZE), decoder_target.reshape(-1))

        scaler.scale(loss).backward()
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(model.parameters(), 5.0)
        scaler.step(optimizer)
        scaler.update()

        train_loss += loss.item()

        with torch.no_grad():
            pred_ids = logits.argmax(dim=-1)
            preds    = [decode_tokens(x) for x in pred_ids]

        train_preds.extend(preds)
        train_labels.extend(labels)
        pbar.set_description(f"Epoch {epoch} | Loss: {loss.item():.4f}")

    scheduler.step()

    train_car       = char_accuracy(train_preds, train_labels)
    train_wer_score = wer(train_labels, train_preds) * 100
    avg_train_loss  = train_loss / len(train_loader)

    # --- VALIDATE (greedy for speed) ---
    model.eval()
    val_loss = 0
    val_preds, val_labels = [], []

    with torch.no_grad():
        for images, targets, labels in tqdm(val_loader, desc="Validating"):
            images, targets = images.to(DEVICE), targets.to(DEVICE)
            decoder_input  = targets[:, :-1]
            decoder_target = targets[:, 1:]

            logits    = model(images, decoder_input)
            loss      = criterion(logits.reshape(-1, VOCAB_SIZE), decoder_target.reshape(-1))
            val_loss += loss.item()

            pred_ids = model.predict(images, use_beam=False)
            preds    = [decode_tokens(x) for x in pred_ids]

            val_preds.extend(preds)
            val_labels.extend(labels)

    val_car       = char_accuracy(val_preds, val_labels)
    val_wer_score = wer(val_labels, val_preds) * 100
    avg_val_loss  = val_loss / len(val_loader)

    print("\n" + "="*60)
    print(f"EPOCH {epoch}  |  LR: {scheduler.get_last_lr()[0]:.2e}")
    print(f"TRAIN LOSS: {avg_train_loss:.4f}  |  VAL LOSS: {avg_val_loss:.4f}")
    print(f"TRAIN CAR:  {train_car:.2f}%      |  VAL CAR:  {val_car:.2f}%")
    print(f"TRAIN WER:  {train_wer_score:.2f}%      |  VAL WER:  {val_wer_score:.2f}%")
    print("="*60)

    if avg_val_loss < best_val_loss:
        best_val_loss    = avg_val_loss
        patience_counter = 0
        torch.save({
            "model":    model.state_dict(),
            "epoch":    epoch,
            "val_loss": avg_val_loss,
        }, CHECKPOINT_PATH)
        print(">>> Checkpoint saved (val loss improved).")
    else:
        patience_counter += 1
        print(f">>> No improvement. Patience: {patience_counter}/{PATIENCE}")

    if patience_counter >= PATIENCE:
        print(f"\n[EARLY STOP] Patience exhausted at epoch {epoch}.")
        break


# ============================================================
# FINAL EVALUATION (beam search)
# ============================================================
print("\nLoading best checkpoint for final evaluation...")
if os.path.exists(CHECKPOINT_PATH):
    checkpoint = torch.load(CHECKPOINT_PATH)
    model.load_state_dict(checkpoint["model"])
    print(f"Loaded checkpoint from epoch {checkpoint['epoch']}")

model.eval()
test_preds, test_labels = [], []

with torch.no_grad():
    for images, _, labels in tqdm(test_loader, desc="Testing with beam search"):
        images       = images.to(DEVICE)
        beam_results = model.predict(images, use_beam=True)
        preds        = decode_beam_results(beam_results)
        test_preds.extend(preds)
        test_labels.extend(labels)

test_car       = char_accuracy(test_preds, test_labels)
test_wer_score = wer(test_labels, test_preds) * 100

print("\n" + "#"*40)
print("FINAL TEST METRICS")
print(f"TEST CAR (Character Accuracy Rate) : {test_car:.2f}%")
print(f"TEST WER (Word Error Rate)         : {test_wer_score:.2f}%")
print("#"*40)


# ============================================================
# SAMPLE PREVIEW
# ============================================================
print("\nSAMPLE PREDICTIONS:\n")
for i in range(min(10, len(test_labels))):
    print(f"[{i+1:02d}] TRUE : {test_labels[i]}")
    print(f"     PRED : {test_preds[i]}")
    print("-" * 50)
```

```python
# new :
#     strong aug
#     label smoothing = 0.1
#     droput=0.3
#     weight_decay=1e-4


import os
import cv2
import time
import random
import unicodedata
import numpy as np
from PIL import Image
from tqdm import tqdm
from sklearn.model_selection import train_test_split

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.optim import Adam
from torch.optim.lr_scheduler import CosineAnnealingLR
from torch.amp import autocast, GradScaler

import timm
from jiwer import wer

try:
    import albumentations as A
    HAS_ALBUMENTATIONS = True
except ImportError:
    HAS_ALBUMENTATIONS = False
    print("albumentations not installed. Using basic augmentation.")


# ============================================================
# CONFIG
# ============================================================
ROOT_DIR    = "/home/mca/Shweta/handwritten text detection/dataset/iam/iam_words"
ORIGINAL_GT = os.path.join(ROOT_DIR, "words.txt")

IMG_HEIGHT = 32
IMG_WIDTH  = 192

BATCH_SIZE = 16
LR         = 5e-5

MAX_EPOCHS = 100
PATIENCE   = 15

MAX_SEQ_LEN = 30
BEAM_WIDTH  = 5
LLRD_DECAY  = 0.75

DEVICE     = "cuda" if torch.cuda.is_available() else "cpu"
NUM_WORKERS = 4
SEED        = 42
USE_AMP      = True

CHECKPOINT_PATH = "best_iam_hvlt_2.pt"


# ============================================================
# SEED
# ============================================================
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

set_seed(SEED)


# ============================================================
# IAM VOCABULARY
# ============================================================
IAM_CHARS = (
    "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
    "0123456789"
    ".,!?()[]{}:;-_'/\\&%$#@+*=\"\u201e\u201c "
)

SPECIAL_TOKENS = ["<PAD>", "<SOS>", "<EOS>", "<UNK>"]
ALL_TOKENS     = SPECIAL_TOKENS + list(IAM_CHARS)

char2idx  = {c: i for i, c in enumerate(ALL_TOKENS)}
idx2char  = {i: c for c, i in char2idx.items()}
VOCAB_SIZE = len(ALL_TOKENS)

PAD_IDX = char2idx["<PAD>"]
SOS_IDX = char2idx["<SOS>"]
EOS_IDX = char2idx["<EOS>"]
UNK_IDX = char2idx["<UNK>"]

print("VOCAB SIZE:", VOCAB_SIZE)


# ============================================================
# DATA LOADING
# ============================================================
samples         = []
missing_count  = 0
corrupted_count = 0

print("Parsing IAM metadata...")
with open(ORIGINAL_GT, "r", encoding="utf-8", errors="ignore") as f:
    for line in f:
        line = line.strip()
        if not line or line.startswith("#"):
            continue

        parts = line.split()
        if len(parts) < 9 or parts[1] == "err":
            continue

        word_id            = parts[0]
        text_transcription = " ".join(parts[8:])

        id_splits   = word_id.split("-")
        f1          = id_splits[0]
        f2          = f"{id_splits[0]}-{id_splits[1]}"
        rel_img_path = f"words/{f1}/{f2}/{word_id}.png"
        full_img_path = os.path.join(ROOT_DIR, rel_img_path)

        if not os.path.exists(full_img_path):
            missing_count += 1
            continue

        try:
            with Image.open(full_img_path) as img:
                img.verify()
            text_transcription = unicodedata.normalize("NFC", text_transcription.strip())
            samples.append((full_img_path, text_transcription))
        except Exception:
            corrupted_count += 1
            continue

print(f"TOTAL SAMPLES    : {len(samples)}")
print(f"MISSING IMAGES   : {missing_count}")
print(f"CORRUPTED IMAGES : {corrupted_count}")

if len(samples) == 0:
    raise ValueError("No samples found. Check ROOT_DIR path.")


# ============================================================
# DATA SPLITTING
# ============================================================
train_samples, temp_samples = train_test_split(samples, test_size=0.15, random_state=SEED)
val_samples, test_samples   = train_test_split(temp_samples, test_size=0.5, random_state=SEED)

print(f"TRAIN : {len(train_samples)}")
print(f"VAL   : {len(val_samples)}")
print(f"TEST  : {len(test_samples)}")


# ============================================================
# ALBUMENTATIONS AUGMENTATION ENGINE
# ============================================================
if HAS_ALBUMENTATIONS:
    strong_aug = A.Compose([
        A.ElasticTransform(alpha=30, sigma=5, p=0.5),
        A.Rotate(limit=5, p=0.5),
        A.GaussNoise(var_limit=(5, 30), p=0.4),
        A.RandomBrightnessContrast(brightness_limit=0.25, contrast_limit=0.25, p=0.6),
        A.GaussianBlur(blur_limit=(3, 5), p=0.3),
        A.Morphological(scale=(1, 2), operation='dilation', p=0.2),
        A.Morphological(scale=(1, 2), operation='erosion',  p=0.2),
    ])
else:
    strong_aug = None


def preprocess_image(img_path, augment=False):
    img = cv2.imread(img_path)
    if img is None:
        img = np.ones((IMG_HEIGHT, IMG_WIDTH, 3), dtype=np.uint8) * 255
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    if augment:
        if strong_aug is not None:
            img = strong_aug(image=img)["image"]
        else:
            if random.random() < 0.5:
                alpha = random.uniform(0.8, 1.2)
                beta  = random.randint(-15, 15)
                img   = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
            if random.random() < 0.3:
                k   = random.choice([3, 5])
                img = cv2.GaussianBlur(img, (k, k), 0)
            if random.random() < 0.4:
                angle = random.uniform(-3, 3)
                M     = cv2.getRotationMatrix2D((img.shape[1] // 2, img.shape[0] // 2), angle, 1.0)
                img = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]), borderValue=(255, 255, 255))

    h, w   = img.shape[:2]
    scale  = IMG_HEIGHT / h
    new_w  = max(1, int(w * scale))
    img    = cv2.resize(img, (new_w, IMG_HEIGHT))

    if new_w < IMG_WIDTH:
        pad = np.ones((IMG_HEIGHT, IMG_WIDTH - new_w, 3), dtype=np.uint8) * 255
        img = np.concatenate([img, pad], axis=1)
    else:
        img = cv2.resize(img, (IMG_WIDTH, IMG_HEIGHT))

    img = img.astype(np.float32) / 255.0
    img = (img - 0.5) / 0.5
    img = np.transpose(img, (2, 0, 1))

    return torch.tensor(img, dtype=torch.float32)


# ============================================================
# TEXT UTILS
# ============================================================
def encode_text(text):
    tokens = [SOS_IDX]
    for ch in text:
        tokens.append(char2idx.get(ch, UNK_IDX))
    tokens.append(EOS_IDX)

    if len(tokens) < MAX_SEQ_LEN:
        tokens += [PAD_IDX] * (MAX_SEQ_LEN - len(tokens))
    else:
        tokens    = tokens[:MAX_SEQ_LEN]
        tokens[-1] = EOS_IDX

    return torch.tensor(tokens, dtype=torch.long)


def decode_tokens(tokens):
    chars = []
    for t in tokens:
        t = int(t)
        if t == EOS_IDX: break
        if t in [PAD_IDX, SOS_IDX]: continue
        chars.append(idx2char.get(t, ""))
    return "".join(chars)


# ============================================================
# DATASET
# ============================================================
class IAMWordDataset(Dataset):
    def __init__(self, dataset_samples, augment=False):
        self.samples = dataset_samples
        self.augment = augment

    def __len__(self): return len(self.samples)

    def __getitem__(self, idx):
        img_path, text = self.samples[idx]
        image  = preprocess_image(img_path, augment=self.augment)
        tokens = encode_text(text)
        return image, tokens, text


train_loader = DataLoader(IAMWordDataset(train_samples, augment=True), batch_size=BATCH_SIZE, shuffle=True, num_workers=NUM_WORKERS, pin_memory=True)
val_loader = DataLoader(IAMWordDataset(val_samples, augment=False), batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)
test_loader = DataLoader(IAMWordDataset(test_samples, augment=False), batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)


# ============================================================
# POSITIONAL BRIDGE
# ============================================================
class PositionalBridge(nn.Module):
    def __init__(self, in_channels=1024, d_model=768, vis_seq_len=192):
        super().__init__()
        self.pool      = nn.AdaptiveAvgPool2d((1, vis_seq_len))
        self.proj      = nn.Linear(in_channels, d_model)
        self.pos_embed = nn.Parameter(torch.randn(1, vis_seq_len, d_model) * 0.02)

    def forward(self, x):
        x = x.permute(0, 3, 1, 2)
        x = self.pool(x).squeeze(2).permute(0, 2, 1)
        return self.proj(x) + self.pos_embed


# ============================================================
# DECODER ENGINE (FIXED DROPOUT AND EMBEDDINGS)
# ============================================================
class IAMDecoder(nn.Module):
    def __init__(self, vocab_size, d_model=768, n_heads=12, n_layers=6):
        super().__init__()
        self.token_embed = nn.Embedding(vocab_size, d_model, padding_idx=PAD_IDX)
        self.pos_embed   = nn.Embedding(MAX_SEQ_LEN, d_model)
        
        # FIX 1 (A): Added Token Embedding Regularization Dropout
        self.embed_dropout = nn.Dropout(0.2)

        # FIX 1 (B): Increased attention dropout to 0.3 to prevent memorization
        decoder_layer = nn.TransformerDecoderLayer(
            d_model=d_model, nhead=n_heads, dim_feedforward=3072,
            dropout=0.3, batch_first=True
        )
        self.decoder     = nn.TransformerDecoder(decoder_layer, num_layers=n_layers)
        self.output_proj = nn.Linear(d_model, vocab_size)
        print("Initialized Character-Level Token Embedding Matrix.")

    def forward(self, memory, tgt_tokens):
        B, T = tgt_tokens.shape
        pos  = torch.arange(T, device=tgt_tokens.device).unsqueeze(0).expand(B, -1)
        
        # FIX 1 (C): Flowing embeddings through token dropout layer
        embeddings = self.token_embed(tgt_tokens) + self.pos_embed(pos)
        x = self.embed_dropout(embeddings)

        tgt_mask             = torch.triu(torch.ones(T, T, device=tgt_tokens.device), diagonal=1).bool()
        tgt_key_padding_mask = (tgt_tokens == PAD_IDX)

        out = self.decoder(tgt=x, memory=memory, tgt_mask=tgt_mask, tgt_key_padding_mask=tgt_key_padding_mask)
        return self.output_proj(out)

    @torch.no_grad()
    def beam_decode(self, memory, beam_width=BEAM_WIDTH, max_len=MAX_SEQ_LEN):
        B       = memory.shape[0]
        results = []

        for b in range(B):
            mem       = memory[b:b+1]
            beams     = [(0.0, [SOS_IDX])]
            completed = []

            for _ in range(max_len):
                candidates = []
                for score, tokens in beams:
                    if tokens[-1] == EOS_IDX:
                        completed.append((score, tokens))
                        continue

                    t     = torch.tensor([tokens], dtype=torch.long, device=memory.device)
                    logit = self.forward(mem, t)
                    log_p = F.log_softmax(logit[0, -1], dim=-1)

                    topk_scores, topk_ids = log_p.topk(beam_width)
                    for s, idx in zip(topk_scores.tolist(), topk_ids.tolist()):
                        candidates.append((score + s, tokens + [idx]))

                if not candidates: break

                candidates.sort(key=lambda x: x[0], reverse=True)
                beams = candidates[:beam_width]

                if all(t[-1] == EOS_IDX for _, t in beams):
                    completed.extend(beams)
                    break

            if not completed: completed = beams
            best_tokens = max(completed, key=lambda x: x[0])[1]
            results.append(torch.tensor(best_tokens[1:], dtype=torch.long))

        return results

    @torch.no_grad()
    def greedy_decode(self, memory, max_len=MAX_SEQ_LEN):
        B         = memory.shape[0]
        generated = torch.full((B, 1), SOS_IDX, device=memory.device, dtype=torch.long)
        finished  = torch.zeros(B, dtype=torch.bool, device=memory.device)

        for _ in range(max_len):
            logits     = self.forward(memory, generated)
            next_token = logits[:, -1].argmax(dim=-1)
            next_token = next_token.masked_fill(finished, PAD_IDX)

            generated = torch.cat([generated, next_token.unsqueeze(1)], dim=1)
            finished |= (next_token == EOS_IDX)
            if finished.all(): break

        return generated[:, 1:]


# ============================================================
# MODEL WRAPPER
# ============================================================
class HVLT(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = timm.create_model(
            "swin_base_patch4_window7_224", pretrained=True, features_only=True,
            img_size=(IMG_HEIGHT, IMG_WIDTH), strict_img_size=False,
        )
        self.bridge  = PositionalBridge(in_channels=1024, d_model=768, vis_seq_len=192)
        self.decoder = IAMDecoder(vocab_size=VOCAB_SIZE)

    def forward(self, images, tgt_tokens):
        feats  = self.encoder(images)[-1]
        memory = self.bridge(feats)
        return self.decoder(memory, tgt_tokens)

    def _encode(self, images):
        feats  = self.encoder(images)[-1]
        return self.bridge(feats)

    @torch.no_grad()
    def predict(self, images, use_beam=True):
        memory = self._encode(images)
        if use_beam: return self.decoder.beam_decode(memory)
        return self.decoder.greedy_decode(memory)


# ============================================================
# LOSS & METRICS
# ============================================================
criterion = nn.CrossEntropyLoss(ignore_index=PAD_IDX, label_smoothing=0.1)


def char_accuracy(preds, labels):
    correct, total = 0, 0
    for p, l in zip(preds, labels):
        n = min(len(p), len(l))
        for i in range(n):
            if p[i] == l[i]: correct += 1
        total += max(len(p), len(l))
    return 100.0 * correct / max(total, 1)


def decode_beam_results(beam_results):
    return [decode_tokens(t.tolist()) for t in beam_results]


# ============================================================
# LAYERWISE LEARNING RATE DECAY STRATIFICATION
# ============================================================
def get_llrd_param_groups(model, base_lr, decay):
    decoder_bridge_params = []
    stage4_params = []
    stage3_params = []
    stage2_params = []
    stage1_embed_params = []

    for name, param in model.named_parameters():
        if not param.requires_grad: continue

        if name.startswith("decoder.") or name.startswith("bridge."):
            decoder_bridge_params.append(param)
        elif "encoder.layers.3" in name or "encoder.norm3" in name:
            stage4_params.append(param)
        elif "encoder.layers.2" in name or "encoder.norm2" in name:
            stage3_params.append(param)
        elif "encoder.layers.1" in name or "encoder.norm1" in name:
            stage2_params.append(param)
        else:
            stage1_embed_params.append(param)

    param_groups = [
        {"params": decoder_bridge_params, "lr": base_lr},
        {"params": stage4_params,         "lr": base_lr * decay},
        {"params": stage3_params,         "lr": base_lr * decay**2},
        {"params": stage2_params,         "lr": base_lr * decay**3},
        {"params": stage1_embed_params,   "lr": base_lr * decay**4},
    ]
    return param_groups


# ============================================================
# INSTANTIATE ENGINES
# ============================================================
model = HVLT().to(DEVICE)

print("\nBuilding LLRD optimizer with structural weight decay regularization:")
param_groups = get_llrd_param_groups(model, LR, LLRD_DECAY)
optimizer = Adam(param_groups, weight_decay=1e-4)

scheduler = CosineAnnealingLR(optimizer, T_max=MAX_EPOCHS, eta_min=1e-7)
scaler = GradScaler("cuda", enabled=USE_AMP)

print("TOTAL PARAMS:", sum(p.numel() for p in model.parameters()) / 1e6, "M")


# ============================================================
# TRAINING LOOP
# ============================================================
best_val_loss    = float('inf')
patience_counter = 0

for epoch in range(1, MAX_EPOCHS + 1):
    model.train()
    train_loss = 0
    train_preds, train_labels = [], []

    pbar = tqdm(train_loader)
    for images, targets, labels in pbar:
        images, targets = images.to(DEVICE), targets.to(DEVICE)
        decoder_input  = targets[:, :-1]
        decoder_target = targets[:, 1:]

        optimizer.zero_grad()
        with autocast("cuda", enabled=USE_AMP):
            logits = model(images, decoder_input)
            loss   = criterion(logits.reshape(-1, VOCAB_SIZE), decoder_target.reshape(-1))

        scaler.scale(loss).backward()
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(model.parameters(), 5.0)
        scaler.step(optimizer)
        scaler.update()

        train_loss += loss.item()

        with torch.no_grad():
            pred_ids = logits.argmax(dim=-1)
            preds    = [decode_tokens(x) for x in pred_ids]

        train_preds.extend(preds)
        train_labels.extend(labels)
        pbar.set_description(f"Epoch {epoch} | Loss: {loss.item():.4f}")

    scheduler.step()

    train_car       = char_accuracy(train_preds, train_labels)
    train_wer_score = wer(train_labels, train_preds) * 100
    avg_train_loss  = train_loss / len(train_loader)

    # --- VALIDATE ---
    model.eval()
    val_loss = 0
    val_preds, val_labels = [], []

    with torch.no_grad():
        for images, targets, labels in tqdm(val_loader, desc="Validating"):
            images, targets = images.to(DEVICE), targets.to(DEVICE)
            decoder_input  = targets[:, :-1]
            decoder_target = targets[:, 1:]

            logits    = model(images, decoder_input)
            loss      = criterion(logits.reshape(-1, VOCAB_SIZE), decoder_target.reshape(-1))
            val_loss += loss.item()

            pred_ids = model.predict(images, use_beam=False)
            preds    = [decode_tokens(x) for x in pred_ids]

            val_preds.extend(preds)
            val_labels.extend(labels)

    val_car       = char_accuracy(val_preds, val_labels)
    val_wer_score = wer(val_labels, val_preds) * 100
    avg_val_loss  = val_loss / len(val_loader)

    print("\n" + "="*60)
    print(f"EPOCH {epoch}  |  LR: {scheduler.get_last_lr()[0]:.2e}")
    print(f"TRAIN LOSS: {avg_train_loss:.4f}  |  VAL LOSS: {avg_val_loss:.4f}")
    print(f"TRAIN CAR:  {train_car:.2f}%      |  VAL CAR:  {val_car:.2f}%")
    print(f"TRAIN WER:  {train_wer_score:.2f}%      |  VAL WER:  {val_wer_score:.2f}%")
    print("="*60)

    if avg_val_loss < best_val_loss:
        best_val_loss    = avg_val_loss
        patience_counter = 0
        torch.save({"model": model.state_dict(), "epoch": epoch, "val_loss": avg_val_loss}, CHECKPOINT_PATH)
        print(">>> Checkpoint saved (val loss improved).")
    else:
        patience_counter += 1
        print(f">>> No improvement. Patience: {patience_counter}/{PATIENCE}")

    if patience_counter >= PATIENCE:
        print(f"\n[EARLY STOP] Patience exhausted at epoch {epoch}.")
        break


# ============================================================
# FINAL EVALUATION (BEAM SEARCH)
# ============================================================
print("\nLoading best checkpoint for final evaluation...")
if os.path.exists(CHECKPOINT_PATH):
    checkpoint = torch.load(CHECKPOINT_PATH)
    model.load_state_dict(checkpoint["model"])
    print(f"Loaded checkpoint from epoch {checkpoint['epoch']}")

model.eval()
test_preds, test_labels = [], []

with torch.no_grad():
    for images, _, labels in tqdm(test_loader, desc="Testing with beam search"):
        images       = images.to(DEVICE)
        beam_results = model.predict(images, use_beam=True)
        preds        = decode_beam_results(beam_results)
        test_preds.extend(preds)
        test_labels.extend(labels)

test_car       = char_accuracy(test_preds, test_labels)
test_wer_score = wer(test_labels, test_preds) * 100

print("\n" + "#"*40)
print("FINAL TEST METRICS")
print(f"TEST CAR (Character Accuracy Rate) : {test_car:.2f}%")
print(f"TEST WER (Word Error Rate)         : {test_wer_score:.2f}%")
print("#"*40)


# ============================================================
# SAMPLE PREVIEW
# ============================================================
print("\nSAMPLE PREDICTIONS:\n")
for i in range(min(10, len(test_labels))):
    print(f"[{i+1:02d}] TRUE : {test_labels[i]}")
    print(f"     PRED : {test_preds[i]}")
    print("-" * 50)
```

## 3 knowledge distillation

```python
# ============================================================
# HVLT IAM - WITH SELF-KNOWLEDGE DISTILLATION
# Uses previous best IAM checkpoint as frozen teacher.
# Student learns from both hard labels (CE) and soft teacher
# distributions (KL divergence).
# ============================================================

import os
import cv2
import random
import unicodedata
import numpy as np
from PIL import Image
from tqdm import tqdm
from sklearn.model_selection import train_test_split

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.optim import Adam
from torch.optim.lr_scheduler import CosineAnnealingLR
from torch.amp import autocast, GradScaler

import timm
from jiwer import wer

try:
    import albumentations as A
    HAS_ALBUMENTATIONS = True
except ImportError:
    HAS_ALBUMENTATIONS = False
    print("albumentations not installed. Using basic augmentation.")


# ============================================================
# CONFIG
# ============================================================
ROOT_DIR    = "/home/mca/Shweta/handwritten text detection/dataset/iam/iam_words"
ORIGINAL_GT = os.path.join(ROOT_DIR, "words.txt")

# ── Distillation settings ──────────────────────────────────
# Point this to your previously trained IAM checkpoint
TEACHER_CHECKPOINT = "best_iam_hvlt.pt"

# If teacher CAR > 70% → alpha=0.7 (trust teacher a lot)
# If teacher CAR 55-70% → alpha=0.4-0.5
# If teacher CAR < 55%  → alpha=0.2 or set USE_DISTILLATION=False
USE_DISTILLATION   = True
DISTILL_ALPHA      = 0.7       # weight of KD loss vs CE loss
DISTILL_TEMPERATURE = 3.0      # higher = softer distributions
# ──────────────────────────────────────────────────────────

IMG_HEIGHT  = 32
IMG_WIDTH   = 192
BATCH_SIZE  = 16
LR          = 5e-5
MAX_EPOCHS  = 100
PATIENCE    = 15
MAX_SEQ_LEN = 30
BEAM_WIDTH  = 5
LLRD_DECAY  = 0.75

DEVICE      = "cuda" if torch.cuda.is_available() else "cpu"
NUM_WORKERS = 4
SEED        = 42
USE_AMP     = True

CHECKPOINT_PATH = "best_iam_hvlt_3.pt"


# ============================================================
# SEED
# ============================================================
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

set_seed(SEED)


# ============================================================
# VOCABULARY
# ============================================================
IAM_CHARS = (
    "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
    "0123456789"
    ".,!?()[]{}:;-_'/\\&%$#@+*=\"\u201e\u201c "
)

SPECIAL_TOKENS = ["<PAD>", "<SOS>", "<EOS>", "<UNK>"]
ALL_TOKENS     = SPECIAL_TOKENS + list(IAM_CHARS)

char2idx  = {c: i for i, c in enumerate(ALL_TOKENS)}
idx2char  = {i: c for c, i in char2idx.items()}
VOCAB_SIZE = len(ALL_TOKENS)

PAD_IDX = char2idx["<PAD>"]
SOS_IDX = char2idx["<SOS>"]
EOS_IDX = char2idx["<EOS>"]
UNK_IDX = char2idx["<UNK>"]

print("VOCAB SIZE:", VOCAB_SIZE)


# ============================================================
# DATA LOADING
# ============================================================
samples         = []
missing_count   = 0
corrupted_count = 0

print("Parsing IAM metadata...")
with open(ORIGINAL_GT, "r", encoding="utf-8", errors="ignore") as f:
    for line in f:
        line = line.strip()
        if not line or line.startswith("#"):
            continue
        parts = line.split()
        if len(parts) < 9 or parts[1] == "err":
            continue

        word_id            = parts[0]
        text_transcription = " ".join(parts[8:])
        id_splits          = word_id.split("-")
        f1                 = id_splits[0]
        f2                 = f"{id_splits[0]}-{id_splits[1]}"
        full_img_path      = os.path.join(ROOT_DIR, f"words/{f1}/{f2}/{word_id}.png")

        if not os.path.exists(full_img_path):
            missing_count += 1
            continue

        try:
            with Image.open(full_img_path) as img:
                img.verify()
            text_transcription = unicodedata.normalize("NFC", text_transcription.strip())
            samples.append((full_img_path, text_transcription))
        except Exception:
            corrupted_count += 1
            continue

print(f"TOTAL SAMPLES    : {len(samples)}")
print(f"MISSING IMAGES   : {missing_count}")
print(f"CORRUPTED IMAGES : {corrupted_count}")

if len(samples) == 0:
    raise ValueError("No samples found. Check ROOT_DIR path.")


# ============================================================
# DATA SPLITTING
# ============================================================
train_samples, temp_samples = train_test_split(samples, test_size=0.15, random_state=SEED)
val_samples, test_samples   = train_test_split(temp_samples, test_size=0.5, random_state=SEED)

print(f"TRAIN : {len(train_samples)}")
print(f"VAL   : {len(val_samples)}")
print(f"TEST  : {len(test_samples)}")


# ============================================================
# AUGMENTATION
# ============================================================
if HAS_ALBUMENTATIONS:
    strong_aug = A.Compose([
        A.ElasticTransform(alpha=30, sigma=5, p=0.5),
        A.Rotate(limit=5, p=0.5),
        A.GaussNoise(var_limit=(5, 30), p=0.4),
        A.RandomBrightnessContrast(brightness_limit=0.25, contrast_limit=0.25, p=0.6),
        A.GaussianBlur(blur_limit=(3, 5), p=0.3),
        A.Morphological(scale=(1, 2), operation='dilation', p=0.2),
        A.Morphological(scale=(1, 2), operation='erosion',  p=0.2),
    ])
else:
    strong_aug = None


def preprocess_image(img_path, augment=False):
    img = cv2.imread(img_path)
    if img is None:
        img = np.ones((IMG_HEIGHT, IMG_WIDTH, 3), dtype=np.uint8) * 255
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    if augment:
        if strong_aug is not None:
            img = strong_aug(image=img)["image"]
        else:
            if random.random() < 0.5:
                img = cv2.convertScaleAbs(img,
                    alpha=random.uniform(0.8, 1.2), beta=random.randint(-15, 15))
            if random.random() < 0.3:
                img = cv2.GaussianBlur(img, (random.choice([3, 5]),) * 2, 0)
            if random.random() < 0.4:
                angle = random.uniform(-3, 3)
                M = cv2.getRotationMatrix2D(
                    (img.shape[1]//2, img.shape[0]//2), angle, 1.0)
                img = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]),
                                     borderValue=(255, 255, 255))

    h, w  = img.shape[:2]
    new_w = max(1, int(w * IMG_HEIGHT / h))
    img   = cv2.resize(img, (new_w, IMG_HEIGHT))

    if new_w < IMG_WIDTH:
        pad = np.ones((IMG_HEIGHT, IMG_WIDTH - new_w, 3), dtype=np.uint8) * 255
        img = np.concatenate([img, pad], axis=1)
    else:
        img = cv2.resize(img, (IMG_WIDTH, IMG_HEIGHT))

    img = img.astype(np.float32) / 255.0
    img = (img - 0.5) / 0.5
    return torch.tensor(np.transpose(img, (2, 0, 1)), dtype=torch.float32)


# ============================================================
# TEXT UTILS
# ============================================================
def encode_text(text):
    tokens = [SOS_IDX]
    for ch in text:
        tokens.append(char2idx.get(ch, UNK_IDX))
    tokens.append(EOS_IDX)
    if len(tokens) < MAX_SEQ_LEN:
        tokens += [PAD_IDX] * (MAX_SEQ_LEN - len(tokens))
    else:
        tokens    = tokens[:MAX_SEQ_LEN]
        tokens[-1] = EOS_IDX
    return torch.tensor(tokens, dtype=torch.long)


def decode_tokens(tokens):
    chars = []
    for t in tokens:
        t = int(t)
        if t == EOS_IDX: break
        if t in [PAD_IDX, SOS_IDX]: continue
        chars.append(idx2char.get(t, ""))
    return "".join(chars)


def decode_beam_results(beam_results):
    return [decode_tokens(t.tolist()) for t in beam_results]


# ============================================================
# DATASET
# ============================================================
class IAMWordDataset(Dataset):
    def __init__(self, dataset_samples, augment=False):
        self.samples = dataset_samples
        self.augment = augment

    def __len__(self): return len(self.samples)

    def __getitem__(self, idx):
        img_path, text = self.samples[idx]
        return preprocess_image(img_path, self.augment), encode_text(text), text


train_loader = DataLoader(IAMWordDataset(train_samples, augment=True),
    batch_size=BATCH_SIZE, shuffle=True,  num_workers=NUM_WORKERS, pin_memory=True)
val_loader   = DataLoader(IAMWordDataset(val_samples,   augment=False),
    batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)
test_loader  = DataLoader(IAMWordDataset(test_samples,  augment=False),
    batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)


# ============================================================
# POSITIONAL BRIDGE
# ============================================================
class PositionalBridge(nn.Module):
    def __init__(self, in_channels=1024, d_model=768, vis_seq_len=192):
        super().__init__()
        self.pool      = nn.AdaptiveAvgPool2d((1, vis_seq_len))
        self.proj      = nn.Linear(in_channels, d_model)
        self.pos_embed = nn.Parameter(torch.randn(1, vis_seq_len, d_model) * 0.02)

    def forward(self, x):
        x = x.permute(0, 3, 1, 2)
        x = self.pool(x).squeeze(2).permute(0, 2, 1)
        return self.proj(x) + self.pos_embed


# ============================================================
# DECODER
# ============================================================
class IAMDecoder(nn.Module):
    def __init__(self, vocab_size, d_model=768, n_heads=12, n_layers=6):
        super().__init__()
        self.token_embed   = nn.Embedding(vocab_size, d_model, padding_idx=PAD_IDX)
        self.pos_embed     = nn.Embedding(MAX_SEQ_LEN, d_model)
        self.embed_dropout = nn.Dropout(0.2)

        decoder_layer = nn.TransformerDecoderLayer(
            d_model=d_model, nhead=n_heads, dim_feedforward=3072,
            dropout=0.3, batch_first=True
        )
        self.decoder     = nn.TransformerDecoder(decoder_layer, num_layers=n_layers)
        self.output_proj = nn.Linear(d_model, vocab_size)
        print("Initialized IAMDecoder.")

    def forward(self, memory, tgt_tokens):
        B, T = tgt_tokens.shape
        pos  = torch.arange(T, device=tgt_tokens.device).unsqueeze(0).expand(B, -1)
        x    = self.embed_dropout(self.token_embed(tgt_tokens) + self.pos_embed(pos))

        tgt_mask             = torch.triu(torch.ones(T, T, device=tgt_tokens.device), diagonal=1).bool()
        tgt_key_padding_mask = (tgt_tokens == PAD_IDX)

        out = self.decoder(tgt=x, memory=memory,
                           tgt_mask=tgt_mask,
                           tgt_key_padding_mask=tgt_key_padding_mask)
        return self.output_proj(out)

    @torch.no_grad()
    def beam_decode(self, memory, beam_width=BEAM_WIDTH, max_len=MAX_SEQ_LEN):
        B, results = memory.shape[0], []
        for b in range(B):
            mem, beams, completed = memory[b:b+1], [(0.0, [SOS_IDX])], []
            for _ in range(max_len):
                candidates = []
                for score, tokens in beams:
                    if tokens[-1] == EOS_IDX:
                        completed.append((score, tokens))
                        continue
                    t     = torch.tensor([tokens], dtype=torch.long, device=memory.device)
                    log_p = F.log_softmax(self.forward(mem, t)[0, -1], dim=-1)
                    for s, idx in zip(*log_p.topk(beam_width)):
                        candidates.append((score + s.item(), tokens + [idx.item()]))
                if not candidates: break
                candidates.sort(key=lambda x: x[0], reverse=True)
                beams = candidates[:beam_width]
                if all(t[-1] == EOS_IDX for _, t in beams):
                    completed.extend(beams)
                    break
            if not completed: completed = beams
            results.append(torch.tensor(
                max(completed, key=lambda x: x[0])[1][1:], dtype=torch.long))
        return results

    @torch.no_grad()
    def greedy_decode(self, memory, max_len=MAX_SEQ_LEN):
        B         = memory.shape[0]
        generated = torch.full((B, 1), SOS_IDX, device=memory.device, dtype=torch.long)
        finished  = torch.zeros(B, dtype=torch.bool, device=memory.device)
        for _ in range(max_len):
            next_token = self.forward(memory, generated)[:, -1].argmax(dim=-1)
            next_token = next_token.masked_fill(finished, PAD_IDX)
            generated  = torch.cat([generated, next_token.unsqueeze(1)], dim=1)
            finished  |= (next_token == EOS_IDX)
            if finished.all(): break
        return generated[:, 1:]


# ============================================================
# HVLT MODEL
# ============================================================
class HVLT(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = timm.create_model(
            "swin_base_patch4_window7_224", pretrained=True, features_only=True,
            img_size=(IMG_HEIGHT, IMG_WIDTH), strict_img_size=False,
        )
        self.bridge  = PositionalBridge(in_channels=1024, d_model=768, vis_seq_len=192)
        self.decoder = IAMDecoder(vocab_size=VOCAB_SIZE)

    def forward(self, images, tgt_tokens):
        return self.decoder(self.bridge(self.encoder(images)[-1]), tgt_tokens)

    def _encode(self, images):
        return self.bridge(self.encoder(images)[-1])

    @torch.no_grad()
    def predict(self, images, use_beam=True):
        memory = self._encode(images)
        return self.decoder.beam_decode(memory) if use_beam \
               else self.decoder.greedy_decode(memory)


# ============================================================
# KNOWLEDGE DISTILLATION LOSS
# ============================================================
class DistillationLoss(nn.Module):
    """
    Combined CE + KL divergence loss for knowledge distillation.

    total_loss = (1 - alpha) * CE(student, hard_labels)
               +      alpha  * KL(student_soft, teacher_soft) * T²

    alpha      : how much to trust teacher vs ground truth
                 0.7 = trust teacher a lot (use when teacher CAR > 70%)
                 0.4 = balanced (teacher CAR 55-70%)
                 0.2 = mostly ground truth (teacher CAR < 55%)

    temperature: controls softness of distributions
                 higher T → smoother, more information shared
                 lower  T → sharper, closer to hard labels
    """
    def __init__(self, temperature=3.0, alpha=0.7):
        super().__init__()
        self.T     = temperature
        self.alpha = alpha
        self.kl    = nn.KLDivLoss(reduction='batchmean')
        self.ce    = nn.CrossEntropyLoss(
            ignore_index=PAD_IDX, label_smoothing=0.1
        )

    def forward(self, student_logits, teacher_logits, hard_targets):
        """
        student_logits  : (B, T, vocab)  — student model output
        teacher_logits  : (B, T, vocab)  — teacher model output (detached)
        hard_targets    : (B, T)         — ground truth token indices
        """
        # ── CE loss against ground truth ──────────────────────
        loss_ce = self.ce(
            student_logits.reshape(-1, VOCAB_SIZE),
            hard_targets.reshape(-1)
        )

        # ── KL loss against teacher soft labels ───────────────
        # Only compute on non-PAD positions to avoid noise
        mask = (hard_targets != PAD_IDX).reshape(-1)

        if mask.sum() == 0:
            return loss_ce

        s_flat = student_logits.reshape(-1, VOCAB_SIZE)[mask]
        t_flat = teacher_logits.reshape(-1, VOCAB_SIZE)[mask]

        # Teacher is already detached (no grad), but be explicit
        t_soft    = F.softmax(t_flat / self.T, dim=-1).detach()
        s_log_soft = F.log_softmax(s_flat / self.T, dim=-1)

        # Multiply by T² to compensate for gradient scaling
        loss_kd = self.kl(s_log_soft, t_soft) * (self.T ** 2)

        total = (1 - self.alpha) * loss_ce + self.alpha * loss_kd
        return total, loss_ce.item(), loss_kd.item()


# ============================================================
# LOAD TEACHER MODEL (frozen)
# ============================================================
teacher = None

if USE_DISTILLATION:
    if os.path.exists(TEACHER_CHECKPOINT):
        print(f"\nLoading teacher from: {TEACHER_CHECKPOINT}")
        teacher = HVLT().to(DEVICE)
        ckpt    = torch.load(TEACHER_CHECKPOINT, map_location=DEVICE)
        teacher.load_state_dict(ckpt["model"])

        # Freeze teacher completely — no gradients, no updates ever
        for param in teacher.parameters():
            param.requires_grad = False
        teacher.eval()

        print(f"Teacher loaded from epoch {ckpt['epoch']} "
              f"(val_loss={ckpt['val_loss']:.4f})")
        print(f"Distillation: alpha={DISTILL_ALPHA}, T={DISTILL_TEMPERATURE}")
        print("Teacher is FROZEN — only student will be updated.\n")
    else:
        print(f"WARNING: Teacher checkpoint not found at {TEACHER_CHECKPOINT}")
        print("Falling back to standard CE loss (no distillation).")
        USE_DISTILLATION = False

# Choose loss function based on whether teacher loaded successfully
if USE_DISTILLATION:
    criterion = DistillationLoss(
        temperature=DISTILL_TEMPERATURE, alpha=DISTILL_ALPHA
    )
else:
    criterion = nn.CrossEntropyLoss(
        ignore_index=PAD_IDX, label_smoothing=0.1
    )
    print("Using standard CrossEntropyLoss.")


# ============================================================
# METRICS
# ============================================================
def char_accuracy(preds, labels):
    correct, total = 0, 0
    for p, l in zip(preds, labels):
        n = min(len(p), len(l))
        correct += sum(p[i] == l[i] for i in range(n))
        total   += max(len(p), len(l))
    return 100.0 * correct / max(total, 1)


# ============================================================
# LLRD OPTIMIZER
# ============================================================
def get_llrd_param_groups(model, base_lr, decay):
    dec_bridge, s4, s3, s2, s1 = [], [], [], [], []

    for name, param in model.named_parameters():
        if not param.requires_grad: continue
        if name.startswith("decoder.") or name.startswith("bridge."):
            dec_bridge.append(param)
        elif "encoder.layers.3" in name or "encoder.norm3" in name:
            s4.append(param)
        elif "encoder.layers.2" in name or "encoder.norm2" in name:
            s3.append(param)
        elif "encoder.layers.1" in name or "encoder.norm1" in name:
            s2.append(param)
        else:
            s1.append(param)

    groups = [
        {"params": dec_bridge, "lr": base_lr},
        {"params": s4,         "lr": base_lr * decay},
        {"params": s3,         "lr": base_lr * decay**2},
        {"params": s2,         "lr": base_lr * decay**3},
        {"params": s1,         "lr": base_lr * decay**4},
    ]

    labels = ["decoder+bridge", "swin stage 4", "swin stage 3",
              "swin stage 2",   "swin stage 1+embed"]
    for lbl, g in zip(labels, groups):
        print(f"  {lbl:25s}  lr={g['lr']:.2e}  ({len(g['params'])} tensors)")

    return groups


# ============================================================
# STUDENT MODEL + OPTIMIZER
# ============================================================
student = HVLT().to(DEVICE)

print("\nBuilding LLRD optimizer:")
param_groups = get_llrd_param_groups(student, LR, LLRD_DECAY)
optimizer    = Adam(param_groups, weight_decay=1e-4)
scheduler    = CosineAnnealingLR(optimizer, T_max=MAX_EPOCHS, eta_min=1e-7)
scaler       = GradScaler("cuda", enabled=USE_AMP)

print(f"\nSTUDENT PARAMS : {sum(p.numel() for p in student.parameters())/1e6:.2f}M")
if teacher is not None:
    print(f"TEACHER PARAMS : {sum(p.numel() for p in teacher.parameters())/1e6:.2f}M (frozen)")


# ============================================================
# TRAINING LOOP
# ============================================================
best_val_loss    = float('inf')
patience_counter = 0

for epoch in range(1, MAX_EPOCHS + 1):

    student.train()
    # Teacher stays eval always — never call teacher.train()
    if teacher is not None:
        teacher.eval()

    train_loss    = 0
    train_ce_sum  = 0   # track CE and KD separately for logging
    train_kd_sum  = 0
    train_preds, train_labels = [], []

    pbar = tqdm(train_loader)
    for images, targets, labels in pbar:
        images, targets = images.to(DEVICE), targets.to(DEVICE)
        decoder_input  = targets[:, :-1]
        decoder_target = targets[:, 1:]

        optimizer.zero_grad()

        with autocast("cuda", enabled=USE_AMP):
            student_logits = student(images, decoder_input)

            if USE_DISTILLATION:
                # Teacher forward pass — no gradients needed
                with torch.no_grad():
                    teacher_logits = teacher(images, decoder_input)

                loss, ce_val, kd_val = criterion(
                    student_logits, teacher_logits, decoder_target
                )
                train_ce_sum += ce_val
                train_kd_sum += kd_val
            else:
                loss = criterion(
                    student_logits.reshape(-1, VOCAB_SIZE),
                    decoder_target.reshape(-1)
                )

        scaler.scale(loss).backward()
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(student.parameters(), 5.0)
        scaler.step(optimizer)
        scaler.update()

        train_loss += loss.item()

        with torch.no_grad():
            preds = [decode_tokens(x) for x in student_logits.argmax(dim=-1)]
        train_preds.extend(preds)
        train_labels.extend(labels)
        pbar.set_description(f"Epoch {epoch} | Loss: {loss.item():.4f}")

    scheduler.step()

    avg_train_loss  = train_loss / len(train_loader)
    train_car       = char_accuracy(train_preds, train_labels)
    train_wer_score = wer(train_labels, train_preds) * 100

    # --- VALIDATE ---
    student.eval()
    val_loss, val_preds, val_labels = 0, [], []

    with torch.no_grad():
        for images, targets, labels in tqdm(val_loader, desc="Validating"):
            images, targets = images.to(DEVICE), targets.to(DEVICE)
            decoder_input  = targets[:, :-1]
            decoder_target = targets[:, 1:]

            student_logits = student(images, decoder_input)

            # Validation always uses plain CE for stable loss tracking
            val_loss += F.cross_entropy(
                student_logits.reshape(-1, VOCAB_SIZE),
                decoder_target.reshape(-1),
                ignore_index=PAD_IDX
            ).item()

            val_preds.extend([decode_tokens(x)
                for x in student.predict(images, use_beam=False)])
            val_labels.extend(labels)

    avg_val_loss  = val_loss / len(val_loader)
    val_car       = char_accuracy(val_preds, val_labels)
    val_wer_score = wer(val_labels, val_preds) * 100

    print("\n" + "="*60)
    print(f"EPOCH {epoch}  |  LR: {scheduler.get_last_lr()[0]:.2e}")
    print(f"TRAIN LOSS: {avg_train_loss:.4f}  |  VAL LOSS: {avg_val_loss:.4f}")
    if USE_DISTILLATION:
        avg_ce = train_ce_sum / len(train_loader)
        avg_kd = train_kd_sum / len(train_loader)
        print(f"  CE component: {avg_ce:.4f}  |  KD component: {avg_kd:.4f}")
    print(f"TRAIN CAR:  {train_car:.2f}%      |  VAL CAR:  {val_car:.2f}%")
    print(f"TRAIN WER:  {train_wer_score:.2f}%      |  VAL WER:  {val_wer_score:.2f}%")
    print("="*60)

    if avg_val_loss < best_val_loss:
        best_val_loss, patience_counter = avg_val_loss, 0
        torch.save({"model": student.state_dict(), "epoch": epoch,
                    "val_loss": avg_val_loss}, CHECKPOINT_PATH)
        print(">>> Checkpoint saved.")
    else:
        patience_counter += 1
        print(f">>> No improvement. Patience: {patience_counter}/{PATIENCE}")

    if patience_counter >= PATIENCE:
        print(f"\n[EARLY STOP] Epoch {epoch}.")
        break


# ============================================================
# FINAL EVALUATION
# ============================================================
print("\nLoading best student checkpoint...")
if os.path.exists(CHECKPOINT_PATH):
    ckpt = torch.load(CHECKPOINT_PATH)
    student.load_state_dict(ckpt["model"])
    print(f"Loaded from epoch {ckpt['epoch']}")

student.eval()
test_preds, test_labels = [], []

with torch.no_grad():
    for images, _, labels in tqdm(test_loader, desc="Testing with beam search"):
        test_preds.extend(decode_beam_results(
            student.predict(images.to(DEVICE), use_beam=True)))
        test_labels.extend(labels)

test_car       = char_accuracy(test_preds, test_labels)
test_wer_score = wer(test_labels, test_preds) * 100

print("\n" + "#"*40)
print("FINAL TEST METRICS (Student after Distillation)")
print(f"TEST CAR : {test_car:.2f}%")
print(f"TEST WER : {test_wer_score:.2f}%")
print("#"*40)

print("\nSAMPLE PREDICTIONS:\n")
for i in range(min(10, len(test_labels))):
    print(f"[{i+1:02d}] TRUE : {test_labels[i]}")
    print(f"     PRED : {test_preds[i]}")
    print("-" * 50)
```

## hyperparameter tuning

```python
# ============================================================
# BAYESIAN HYPERPARAMETER TUNING FOR HVLT IAM
# Uses Optuna — the standard library for Bayesian optimization
# in deep learning research.
#
# pip install optuna optuna-dashboard
#
# STRATEGY:
#    - Each "trial" trains for MAX_TUNE_EPOCHS (short run)
#    - Optuna learns which hyperparams give best val CAR
#    - After N_TRIALS, take best params and do full training
#
# RUNTIME ESTIMATE:
#    N_TRIALS=20, MAX_TUNE_EPOCHS=15 → ~20 × 35min = ~12hrs
#    Run overnight. Reduce N_TRIALS=10 if time is short.
# ============================================================

import os
import cv2
import random
import unicodedata
import numpy as np
from PIL import Image
from tqdm import tqdm
from sklearn.model_selection import train_test_split

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
from torch.optim import Adam
from torch.optim.lr_scheduler import CosineAnnealingLR
from torch.amp import autocast, GradScaler

import timm
import optuna
from optuna.samplers import TPESampler    # Tree-structured Parzen Estimator = Bayesian
from jiwer import wer

try:
    import albumentations as A
    HAS_ALBUMENTATIONS = True
except ImportError:
    HAS_ALBUNCH_ALBUMENTATIONS = False

# ============================================================
# FIXED CONFIG (not tuned)
# ============================================================
ROOT_DIR    = "/home/mca/Shweta/handwritten text detection/dataset/iam/iam_words"
ORIGINAL_GT = os.path.join(ROOT_DIR, "words.txt")

IMG_HEIGHT  = 32
IMG_WIDTH   = 192
MAX_SEQ_LEN = 30
BEAM_WIDTH  = 5

DEVICE      = "cuda" if torch.cuda.is_available() else "cpu"
NUM_WORKERS = 4
SEED        = 42
USE_AMP     = True

# Tuning budget — adjust based on your time
N_TRIALS        = 20    # number of Optuna trials
MAX_TUNE_EPOCHS = 15    # short runs during search (increase to 20 if GPU allows)
PATIENCE_TUNE   = 5     # early stopping during tuning runs

# After tuning — final full training with best params
MAX_FINAL_EPOCHS = 100
PATIENCE_FINAL   = 15

STUDY_NAME      = "hvlt_iam_tuning"
STUDY_TXT_LOG   = "hvlt_iam_optuna_progress.txt"   # saves progress to a text file instead of .db
BEST_PARAMS_PATH = "best_hyperparams.txt"
FINAL_CHECKPOINT = "best_iam_hvlt_tuned.pt"


# ============================================================
# VOCABULARY
# ============================================================
IAM_CHARS = (
    "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ"
    "0123456789"
    ".,!?()[]{}:;-_'/\\&%$#@+*=\"\u201e\u201c "
)
SPECIAL_TOKENS = ["<PAD>", "<SOS>", "<EOS>", "<UNK>"]
ALL_TOKENS     = SPECIAL_TOKENS + list(IAM_CHARS)
char2idx       = {c: i for i, c in enumerate(ALL_TOKENS)}
idx2char       = {i: c for c, i in char2idx.items()}
VOCAB_SIZE     = len(ALL_TOKENS)
PAD_IDX = char2idx["<PAD>"]
SOS_IDX = char2idx["<SOS>"]
EOS_IDX = char2idx["<EOS>"]
UNK_IDX = char2idx["<UNK>"]

print(f"VOCAB SIZE: {VOCAB_SIZE}")


# ============================================================
# SEED
# ============================================================
def set_seed(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

set_seed(SEED)


# ============================================================
# DATA LOADING (done once, reused across all trials)
# ============================================================
def load_iam_samples():
    samples         = []
    missing_count   = 0
    corrupted_count = 0

    with open(ORIGINAL_GT, "r", encoding="utf-8", errors="ignore") as f:
        for line in f:
            line = line.strip()
            if not line or line.startswith("#"):
                continue
            parts = line.split()
            if len(parts) < 9 or parts[1] == "err":
                continue
            word_id            = parts[0]
            text_transcription = " ".join(parts[8:])
            id_splits          = word_id.split("-")
            f1                 = id_splits[0]
            f2                 = f"{id_splits[0]}-{id_splits[1]}"
            full_img_path      = os.path.join(ROOT_DIR, f"words/{f1}/{f2}/{word_id}.png")

            if not os.path.exists(full_img_path):
                missing_count += 1
                continue
            try:
                with Image.open(full_img_path) as img:
                    img.verify()
                samples.append((full_img_path,
                    unicodedata.normalize("NFC", text_transcription.strip())))
            except Exception:
                corrupted_count += 1

    print(f"Loaded {len(samples)} samples ({missing_count} missing, {corrupted_count} corrupted)")
    return samples


print("Loading IAM dataset (once for all trials)...")
ALL_SAMPLES = load_iam_samples()

train_samples, temp = train_test_split(ALL_SAMPLES, test_size=0.15, random_state=SEED)
val_samples, test_samples = train_test_split(temp, test_size=0.5, random_state=SEED)

print(f"TRAIN:{len(train_samples)}  VAL:{len(val_samples)}  TEST:{len(test_samples)}")


# ============================================================
# PREPROCESSING & DATASET
# ============================================================
def preprocess_image(img_path, augment=False, aug_pipeline=None):
    img = cv2.imread(img_path)
    if img is None:
        img = np.ones((IMG_HEIGHT, IMG_WIDTH, 3), dtype=np.uint8) * 255
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    if augment and aug_pipeline is not None:
        img = aug_pipeline(image=img)["image"]

    h, w  = img.shape[:2]
    new_w = max(1, int(w * IMG_HEIGHT / h))
    img   = cv2.resize(img, (new_w, IMG_HEIGHT))
    if new_w < IMG_WIDTH:
        pad = np.ones((IMG_HEIGHT, IMG_WIDTH - new_w, 3), dtype=np.uint8) * 255
        img = np.concatenate([img, pad], axis=1)
    else:
        img = cv2.resize(img, (IMG_WIDTH, IMG_HEIGHT))

    img = img.astype(np.float32) / 255.0
    img = (img - 0.5) / 0.5
    return torch.tensor(np.transpose(img, (2, 0, 1)), dtype=torch.float32)


def build_aug_pipeline(elastic_alpha, elastic_sigma, rotate_limit,
                       noise_var, brightness_limit):
    """Build albumentations pipeline from tunable params."""
    if not HAS_ALBUMENTATIONS:
        return None
    return A.Compose([
        A.ElasticTransform(alpha=elastic_alpha, sigma=elastic_sigma, p=0.5),
        A.Rotate(limit=rotate_limit, p=0.5),
        A.GaussNoise(var_limit=(5, noise_var), p=0.4),
        A.RandomBrightnessContrast(
            brightness_limit=brightness_limit,
            contrast_limit=brightness_limit, p=0.6),
        A.GaussianBlur(blur_limit=(3, 5), p=0.3),
        A.Morphological(scale=(1, 2), operation='dilation', p=0.2),
        A.Morphological(scale=(1, 2), operation='erosion',  p=0.2),
    ])


def encode_text(text):
    tokens = [SOS_IDX] + [char2idx.get(c, UNK_IDX) for c in text] + [EOS_IDX]
    if len(tokens) < MAX_SEQ_LEN:
        tokens += [PAD_IDX] * (MAX_SEQ_LEN - len(tokens))
    else:
        tokens = tokens[:MAX_SEQ_LEN]
        tokens[-1] = EOS_IDX
    return torch.tensor(tokens, dtype=torch.long)


def decode_tokens(tokens):
    chars = []
    for t in tokens:
        t = int(t)
        if t == EOS_IDX: break
        if t in [PAD_IDX, SOS_IDX]: continue
        chars.append(idx2char.get(t, ""))
    return "".join(chars)


class IAMDataset(Dataset):
    def __init__(self, samples, augment=False, aug_pipeline=None):
        self.samples      = samples
        self.augment      = augment
        self.aug_pipeline = aug_pipeline

    def __len__(self): return len(self.samples)

    def __getitem__(self, idx):
        img_path, text = self.samples[idx]
        img = preprocess_image(img_path, self.augment, self.aug_pipeline)
        return img, encode_text(text), text


# ============================================================
# MODEL COMPONENTS
# ============================================================
class PositionalBridge(nn.Module):
    def __init__(self, in_channels=1024, d_model=768, vis_seq_len=192):
        super().__init__()
        self.pool      = nn.AdaptiveAvgPool2d((1, vis_seq_len))
        self.proj      = nn.Linear(in_channels, d_model)
        self.pos_embed = nn.Parameter(torch.randn(1, vis_seq_len, d_model) * 0.02)

    def forward(self, x):
        x = x.permute(0, 3, 1, 2)
        x = self.pool(x).squeeze(2).permute(0, 2, 1)
        return self.proj(x) + self.pos_embed


class IAMDecoder(nn.Module):
    def __init__(self, vocab_size, d_model=768, n_heads=12,
                 n_layers=6, dropout=0.3, embed_dropout=0.2):
        super().__init__()
        self.token_embed   = nn.Embedding(vocab_size, d_model, padding_idx=PAD_IDX)
        self.pos_embed     = nn.Embedding(MAX_SEQ_LEN, d_model)
        self.embed_dropout = nn.Dropout(embed_dropout)
        decoder_layer = nn.TransformerDecoderLayer(
            d_model=d_model, nhead=n_heads, dim_feedforward=3072,
            dropout=dropout, batch_first=True)
        self.decoder     = nn.TransformerDecoder(decoder_layer, num_layers=n_layers)
        self.output_proj = nn.Linear(d_model, vocab_size)

    def forward(self, memory, tgt_tokens):
        B, T = tgt_tokens.shape
        pos  = torch.arange(T, device=tgt_tokens.device).unsqueeze(0).expand(B, -1)
        x    = self.embed_dropout(self.token_embed(tgt_tokens) + self.pos_embed(pos))
        tgt_mask = torch.triu(torch.ones(T, T, device=tgt_tokens.device), diagonal=1).bool()
        out = self.decoder(tgt=x, memory=memory, tgt_mask=tgt_mask,
                           tgt_key_padding_mask=(tgt_tokens == PAD_IDX))
        return self.output_proj(out)

    @torch.no_grad()
    def greedy_decode(self, memory, max_len=MAX_SEQ_LEN):
        B         = memory.shape[0]
        generated = torch.full((B, 1), SOS_IDX, device=memory.device, dtype=torch.long)
        finished  = torch.zeros(B, dtype=torch.bool, device=memory.device)
        for _ in range(max_len):
            next_tok = self.forward(memory, generated)[:, -1].argmax(dim=-1)
            next_tok = next_tok.masked_fill(finished, PAD_IDX)
            generated = torch.cat([generated, next_tok.unsqueeze(1)], dim=1)
            finished |= (next_tok == EOS_IDX)
            if finished.all(): break
        return generated[:, 1:]


class HVLT(nn.Module):
    def __init__(self, dropout=0.3, embed_dropout=0.2):
        super().__init__()
        self.encoder = timm.create_model(
            "swin_base_patch4_window7_224", pretrained=True, features_only=True,
            img_size=(IMG_HEIGHT, IMG_WIDTH), strict_img_size=False)
        self.bridge  = PositionalBridge(in_channels=1024, d_model=768, vis_seq_len=192)
        self.decoder = IAMDecoder(VOCAB_SIZE, dropout=dropout,
                                   embed_dropout=embed_dropout)

    def forward(self, images, tgt_tokens):
        return self.decoder(self.bridge(self.encoder(images)[-1]), tgt_tokens)

    def encode(self, images):
        return self.bridge(self.encoder(images)[-1])

    @torch.no_grad()
    def predict_greedy(self, images):
        return self.decoder.greedy_decode(self.encode(images))


# ============================================================
# METRICS
# ============================================================
def char_accuracy(preds, labels):
    correct, total = 0, 0
    for p, l in zip(preds, labels):
        n = min(len(p), len(l))
        correct += sum(p[i] == l[i] for i in range(n))
        total   += max(len(p), len(l))
    return 100.0 * correct / max(total, 1)


def get_llrd_groups(model, base_lr, decay, weight_decay):
    dec_bridge, s4, s3, s2, s1 = [], [], [], [], []
    for name, param in model.named_parameters():
        if not param.requires_grad: continue
        if name.startswith("decoder.") or name.startswith("bridge."):
            dec_bridge.append(param)
        elif "encoder.layers.3" in name or "encoder.norm3" in name:
            s4.append(param)
        elif "encoder.layers.2" in name or "encoder.norm2" in name:
            s3.append(param)
        elif "encoder.layers.1" in name or "encoder.norm1" in name:
            s2.append(param)
        else:
            s1.append(param)
    return [
        {"params": dec_bridge, "lr": base_lr,            "weight_decay": weight_decay},
        {"params": s4,         "lr": base_lr * decay,    "weight_decay": weight_decay},
        {"params": s3,         "lr": base_lr * decay**2, "weight_decay": weight_decay},
        {"params": s2,         "lr": base_lr * decay**3, "weight_decay": weight_decay},
        {"params": s1,         "lr": base_lr * decay**4, "weight_decay": weight_decay},
    ]


# ============================================================
# OPTUNA OBJECTIVE FUNCTION
# Each call to objective() = one full short training run
# ============================================================
def objective(trial):
    """
    Optuna calls this function repeatedly.
    It suggests hyperparameters, trains for MAX_TUNE_EPOCHS,
    and returns val CAR (higher = better).
    """

    # ── Hyperparameters being searched ────────────────────
    # These are the ranges Optuna will explore:
    lr            = trial.suggest_float("lr",           1e-5,  2e-4, log=True)
    weight_decay  = trial.suggest_float("weight_decay", 1e-5,  1e-3, log=True)
    llrd_decay    = trial.suggest_float("llrd_decay",   0.60,  0.90)
    dropout       = trial.suggest_float("dropout",      0.1,   0.4)
    embed_dropout = trial.suggest_float("embed_dropout",0.1,   0.3)
    label_smooth  = trial.suggest_float("label_smooth", 0.05,  0.20)
    batch_size    = trial.suggest_categorical("batch_size", [8, 16, 32])

    # Augmentation params
    elastic_alpha  = trial.suggest_float("elastic_alpha",  10,  50)
    elastic_sigma  = trial.suggest_float("elastic_sigma",  3,   8)
    rotate_limit   = trial.suggest_int("rotate_limit",     2,   8)
    noise_var      = trial.suggest_int("noise_var",        15,  50)
    brightness_lim = trial.suggest_float("brightness_lim", 0.1, 0.4)
    # ──────────────────────────────────────────────────────

    print(f"\n{'='*50}")
    print(f"TRIAL {trial.number}")
    print(f"  lr={lr:.2e}  wd={weight_decay:.2e}  llrd={llrd_decay:.2f}")
    print(f"  dropout={dropout:.2f}  embed_drop={embed_dropout:.2f}")
    print(f"  label_smooth={label_smooth:.2f}  batch={batch_size}")
    print(f"{'='*50}")

    # Build augmentation pipeline with trial params
    aug_pipeline = build_aug_pipeline(
        elastic_alpha, elastic_sigma, rotate_limit, noise_var, brightness_lim
    )

    # Build dataloaders with trial batch size
    t_loader = DataLoader(
        IAMDataset(train_samples, augment=True, aug_pipeline=aug_pipeline),
        batch_size=batch_size, shuffle=True, num_workers=NUM_WORKERS,
        pin_memory=True, drop_last=True)
    v_loader = DataLoader(
        IAMDataset(val_samples, augment=False),
        batch_size=batch_size, shuffle=False, num_workers=NUM_WORKERS,
        pin_memory=True)

    # Build model with trial dropout
    model     = HVLT(dropout=dropout, embed_dropout=embed_dropout).to(DEVICE)
    criterion = nn.CrossEntropyLoss(ignore_index=PAD_IDX,
                                     label_smoothing=label_smooth)
    groups    = get_llrd_groups(model, lr, llrd_decay, weight_decay)
    optimizer = Adam(groups)
    scheduler = CosineAnnealingLR(optimizer, T_max=MAX_TUNE_EPOCHS, eta_min=1e-7)
    scaler    = GradScaler("cuda", enabled=USE_AMP)

    best_val_car     = 0.0
    best_val_loss    = float('inf')
    patience_counter = 0

    for epoch in range(1, MAX_TUNE_EPOCHS + 1):
        # --- Train ---
        model.train()
        for images, targets, _ in t_loader:
            images, targets = images.to(DEVICE), targets.to(DEVICE)
            optimizer.zero_grad()
            with autocast("cuda", enabled=USE_AMP):
                logits = model(images, targets[:, :-1])
                loss   = criterion(logits.reshape(-1, VOCAB_SIZE),
                                   targets[:, 1:].reshape(-1))
            scaler.scale(loss).backward()
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(model.parameters(), 5.0)
            scaler.step(optimizer)
            scaler.update()
        scheduler.step()

        # --- Validate ---
        model.eval()
        val_loss_sum = 0
        val_preds, val_labels = [], []

        with torch.no_grad():
            for images, targets, labels in v_loader:
                images, targets = images.to(DEVICE), targets.to(DEVICE)
                logits       = model(images, targets[:, :-1])
                val_loss_sum += criterion(logits.reshape(-1, VOCAB_SIZE),
                                          targets[:, 1:].reshape(-1)).item()
                pred_ids = model.predict_greedy(images)
                val_preds.extend([decode_tokens(x) for x in pred_ids])
                val_labels.extend(labels)

        avg_val_loss = val_loss_sum / len(v_loader)
        val_car      = char_accuracy(val_preds, val_labels)

        print(f"  Epoch {epoch:02d} | val_loss={avg_val_loss:.4f} | val_CAR={val_car:.2f}%")

        if val_car > best_val_car:
            best_val_car  = val_car
            best_val_loss = avg_val_loss
            patience_counter = 0
        else:
            patience_counter += 1

        # Report to Optuna for pruning (stops bad trials early)
        trial.report(val_car, epoch)
        if trial.should_prune():
            print(f"  Trial {trial.number} PRUNED at epoch {epoch}")
            raise optuna.exceptions.TrialPruned()

        if patience_counter >= PATIENCE_TUNE:
            print(f"  Early stop at epoch {epoch}")
            break

    print(f"  → Trial {trial.number} best val CAR: {best_val_car:.2f}%")

    # Free GPU memory between trials
    del model, optimizer, scheduler, scaler
    torch.cuda.empty_cache()

    return best_val_car   # Optuna maximizes this


# ============================================================
# TEXT FILE LOGGER CALLBACK FOR OPTUNA
# This logs progress live to a text file after each trial ends
# ============================================================
def txt_logging_callback(study, trial):
    with open(STUDY_TXT_LOG, "a") as log_file:
        log_file.write(f"Trial {trial.number} finished with value (Val CAR): {trial.value:.2f}%\n")
        log_file.write("Parameters:\n")
        for k, v in trial.params.items():
            log_file.write(f"  {k}: {v}\n")
        log_file.write(f"State: {trial.state}\n")
        log_file.write("-" * 40 + "\n")


# ============================================================
# RUN OPTUNA STUDY
# ============================================================
print(f"\nStarting Bayesian hyperparameter search")
print(f"  Trials: {N_TRIALS}")
print(f"  Epochs per trial: {MAX_TUNE_EPOCHS}")
print(f"  Estimated time: ~{N_TRIALS * MAX_TUNE_EPOCHS * 2.5 / 60:.1f} hours")
print(f"  Progress saved to: {STUDY_TXT_LOG}\n")

# Initialize text file with headers
with open(STUDY_TXT_LOG, "w") as f:
    f.write(f"Optuna Study Hyperparameter Tuning Progress Log\n")
    f.write(f"Study Name: {STUDY_NAME}\n")
    f.write("=" * 50 + "\n")

# TPESampler = Tree-structured Parzen Estimator = Bayesian optimization
# MedianPruner = stops trials that are below median performance at each epoch
sampler = TPESampler(seed=SEED, n_startup_trials=5)
pruner  = optuna.pruners.MedianPruner(n_startup_trials=5, n_warmup_steps=3)

# Uses in-memory storage to skip generating any SQLite .db files
study = optuna.create_study(
    study_name   = STUDY_NAME,
    storage      = None,            # None defaults to InMemoryStorage
    direction    = "maximize",    # maximize val CAR
    sampler      = sampler,
    pruner       = pruner
)

# Added the txt_logging_callback here to write trial statistics out instantly
study.optimize(objective, n_trials=N_TRIALS, show_progress_bar=True, callbacks=[txt_logging_callback])


# ============================================================
# REPORT BEST PARAMS
# ============================================================
best = study.best_trial
print("\n" + "#"*50)
print("BEST TRIAL")
print(f"  Val CAR : {best.value:.2f}%")
print(f"  Trial # : {best.number}")
print("\nBEST HYPERPARAMETERS:")
for k, v in best.params.items():
    print(f"  {k:20s} = {v}")
print("#"*50)

# Save best params to file
with open(BEST_PARAMS_PATH, "w") as f:
    f.write(f"Best val CAR: {best.value:.2f}%\n\n")
    for k, v in best.params.items():
        f.write(f"{k} = {v}\n")
print(f"\nBest params saved to {BEST_PARAMS_PATH}")

# Show top 5 trials
print("\nTOP 5 TRIALS:")
trials_df = study.trials_dataframe()
top5 = trials_df.sort_values("value", ascending=False).head(5)
print(top5[["number", "value"] +
           [c for c in top5.columns if c.startswith("params_")]].to_string())


# ============================================================
# FULL TRAINING WITH BEST PARAMS
# ============================================================
print("\n" + "="*60)
print("STARTING FULL TRAINING WITH BEST HYPERPARAMETERS")
print("="*60)

p = best.params  # shorthand

# Rebuild augmentation with best params
best_aug = build_aug_pipeline(
    p["elastic_alpha"], p["elastic_sigma"], p["rotate_limit"],
    p["noise_var"],     p["brightness_lim"]
)

train_loader = DataLoader(
    IAMDataset(train_samples, augment=True, aug_pipeline=best_aug),
    batch_size=p["batch_size"], shuffle=True,
    num_workers=NUM_WORKERS, pin_memory=True)
val_loader = DataLoader(
    IAMDataset(val_samples, augment=False),
    batch_size=p["batch_size"], shuffle=False,
    num_workers=NUM_WORKERS, pin_memory=True)
test_loader = DataLoader(
    IAMDataset(test_samples, augment=False),
    batch_size=p["batch_size"], shuffle=False,
    num_workers=NUM_WORKERS, pin_memory=True)

model     = HVLT(dropout=p["dropout"], embed_dropout=p["embed_dropout"]).to(DEVICE)
criterion = nn.CrossEntropyLoss(ignore_index=PAD_IDX,
                                 label_smoothing=p["label_smooth"])
groups    = get_llrd_groups(model, p["lr"], p["llrd_decay"], p["weight_decay"])
optimizer = Adam(groups)
scheduler = CosineAnnealingLR(optimizer, T_max=MAX_FINAL_EPOCHS, eta_min=1e-7)
scaler    = GradScaler("cuda", enabled=USE_AMP)

best_val_loss    = float('inf')
patience_counter = 0

for epoch in range(1, MAX_FINAL_EPOCHS + 1):
    model.train()
    train_loss, train_preds, train_labels = 0, [], []

    pbar = tqdm(train_loader, desc=f"Epoch {epoch}")
    for images, targets, labels in pbar:
        images, targets = images.to(DEVICE), targets.to(DEVICE)
        optimizer.zero_grad()
        with autocast("cuda", enabled=USE_AMP):
            logits = model(images, targets[:, :-1])
            loss   = criterion(logits.reshape(-1, VOCAB_SIZE),
                               targets[:, 1:].reshape(-1))
        scaler.scale(loss).backward()
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(model.parameters(), 5.0)
        scaler.step(optimizer)
        scaler.update()

        train_loss += loss.item()
        with torch.no_grad():
            train_preds.extend([decode_tokens(x) for x in logits.argmax(dim=-1)])
        train_labels.extend(labels)
        pbar.set_description(f"Epoch {epoch} | loss={loss.item():.4f}")

    scheduler.step()
    train_car       = char_accuracy(train_preds, train_labels)
    train_wer_score = wer(train_labels, train_preds) * 100
    avg_train_loss  = train_loss / len(train_loader)

    model.eval()
    val_loss, val_preds, val_labels = 0, [], []
    with torch.no_grad():
        for images, targets, labels in tqdm(val_loader, desc="Validating"):
            images, targets = images.to(DEVICE), targets.to(DEVICE)
            logits    = model(images, targets[:, :-1])
            val_loss += criterion(logits.reshape(-1, VOCAB_SIZE),
                                   targets[:, 1:].reshape(-1)).item()
            val_preds.extend([decode_tokens(x)
                for x in model.predict_greedy(images)])
            val_labels.extend(labels)

    avg_val_loss  = val_loss / len(val_loader)
    val_car       = char_accuracy(val_preds, val_labels)
    val_wer_score = wer(val_labels, val_preds) * 100

    print(f"\n{'='*60}")
    print(f"EPOCH {epoch}  |  LR: {scheduler.get_last_lr()[0]:.2e}")
    print(f"TRAIN LOSS: {avg_train_loss:.4f}  |  VAL LOSS: {avg_val_loss:.4f}")
    print(f"TRAIN CAR:  {train_car:.2f}%      |  VAL CAR:  {val_car:.2f}%")
    print(f"TRAIN WER:  {train_wer_score:.2f}%      |  VAL WER:  {val_wer_score:.2f}%")
    print("="*60)

    if avg_val_loss < best_val_loss:
        best_val_loss, patience_counter = avg_val_loss, 0
        torch.save({"model": model.state_dict(), "epoch": epoch,
                    "val_loss": avg_val_loss, "best_params": p},
                   FINAL_CHECKPOINT)
        print(">>> Checkpoint saved.")
    else:
        patience_counter += 1
        print(f">>> No improvement. Patience: {patience_counter}/{PATIENCE_FINAL}")
    if patience_counter >= PATIENCE_FINAL:
        print(f"\n[EARLY STOP] Epoch {epoch}")
        break


# ============================================================
# FINAL TEST EVALUATION
# ============================================================
print("\nLoading best checkpoint for final evaluation...")
ckpt = torch.load(FINAL_CHECKPOINT)
model.load_state_dict(ckpt["model"])
model.eval()

test_preds, test_labels = [], []
with torch.no_grad():
    for images, _, labels in tqdm(test_loader, desc="Testing"):
        pred_ids = model.predict_greedy(images.to(DEVICE))
        test_preds.extend([decode_tokens(x) for x in pred_ids])
        test_labels.extend(labels)

test_car       = char_accuracy(test_preds, test_labels)
test_wer_score = wer(test_labels, test_preds) * 100

print("\n" + "#"*50)
print("FINAL TEST METRICS (tuned model)")
print(f"TEST CAR : {test_car:.2f}%")
print(f"TEST WER : {test_wer_score:.2f}%")
print("#"*50)

print("\nSAMPLE PREDICTIONS:")
for i in range(min(10, len(test_labels))):
    print(f"[{i+1:02d}] TRUE : {test_labels[i]}")
    print(f"     PRED : {test_preds[i]}")
    print("-" * 50)
```
